#Overview

A versatile Python toolkit built for data cleaning, exploration, and interactive visualization, especially suited for Google Colab environments. It simplifies common preprocessing workflows such as data sanitization, handling missing values, detecting outliers, and performing advanced statistical relationship analysis.

## Features

* **Smart Data Loading**: Automatically detects and standardizes common missing-value indicators (such as `?`, `N/A`, `NULL`) and attempts to infer and convert columns into appropriate numeric formats.

* **Comprehensive Data Inspection**: Provides quick access to dataset structure, column data types, and summary statistics for both numerical and categorical variables.

* **Automated Data Cleaning**:

  * Impute missing values using mean, median, mode, or fixed-value strategies.
  * Identify and remove duplicate records.
  * Detect and filter outliers using IQR-based methods.
  * Supports interactive removal of selected rows and columns.

* **Advanced Feature Scaling & Encoding**:

  * **Numerical features**: Includes Min-Max scaling, Standard (Z-score) normalization, and Robust scaling for datasets with outliers.
  * **Categorical features**: Supports One-Hot encoding, Ordinal encoding, and uniform encoding approaches.

* **Interactive Data Visualization** (powered by Plotly):

  * Includes violin plots, scatter plots, histograms, and grouped bar charts.
  * Automatically adapts visualization types based on the nature of the data.

* **Advanced Statistical Insights**:

  * Pearson correlation heatmaps for numerical variables.
  * Cramér’s V analysis for categorical variable relationships.
  * Unified association heatmaps combining numeric and categorical features using methods such as point-biserial correlation and ANOVA-based measures.

## Quick Start (Use Cases)

This toolkit is designed primarily for use in **Google Colab**, enabling smooth, interactive workflows for data preprocessing and analysis.

### 1. Data Cleaning & Missing Value Handling

Load your dataset and perform basic cleaning in a single pipeline.

```python
from data_analysis import DataInspector

inspector = DataInspector()

# Step 1: Upload CSV file interactively in Colab
inspector.upload_data()

# Step 2: Fill missing values using median imputation
inspector.handle_missing_values(strategy='median')

# Step 3: Remove duplicate records
inspector.remove_duplicates()
```

### 2. Exploratory Data Analysis (EDA)

Create quick visual insights into your dataset.

```python
# Plot numerical feature distributions (Violin, Scatter, Histogram combined)
inspector.plot_numerical(['Age', 'Salary'])

# Analyze relationships between variables
# (Automatically chooses best plot type: scatter, box, or bar)
inspector.plot_relationship('Department', 'Salary')
```

### 3. Feature Engineering & Data Normalization

Prepare data for machine learning workflows.

```python
# Apply Robust Scaling (effective for outlier-heavy data)
normalized_numeric = inspector.extract_normalized_numeric_data(method='robust')

# Convert categorical variables using One-Hot Encoding
encoded_cat = inspector.extract_normalized_categorical_data(method='onehot')

# Merge processed features into a final training-ready dataset
final_df = inspector.create_normalized_data_df()
```

### 4. Correlation & Association Analysis

Discover relationships across both numerical and categorical variables.

```python
# Generate a combined association heatmap
inspector.plot_all_associations_heatmap()
```

### 5. Custom Visualization Tools

The `PlottingMethods` class allows direct creation of specific chart types. Each function returns a Plotly-compatible output that can be displayed in Colab using `inspector.display_image(result)`.

#### Bar Charts

```python
# Grouped or stacked bar chart visualization
result = plotter.plot_bar_chart(
    x='Department',
    y='Salary',
    color='Gender',
    barmode='group',
    data=my_json_data
)
plotter.display_image(result)
```

#### Pie Charts

```python
# Donut-style pie chart for category distribution
result = plotter.plot_pie_chart(
    names='Category',
    values='Total',
    hole=0.4,
    title='Revenue Split'
)
plotter.display_image(result)
```

#### Histograms

```python
# Histogram with custom bin ranges
result = plotter.plot_histogram(
    x='Age',
    bins=[0, 18, 35, 60, 100],
    title='Age Demographics'
)
plotter.display_image(result)
```



### 6. Implementation Note for Google Colab
To view the output of custom `PlottingMethods`, always use the built-in display helper:
```python
result = inspector.plot_bar_chart(...)
inspector.display_image(result)
```
