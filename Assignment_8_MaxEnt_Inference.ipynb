## **Q1) Statistical Mechanics of the Quantum Oscillator**

## 1. Canonical Partition Function of a Quantum Harmonic Oscillator

The energy eigenvalues of a single quantum harmonic oscillator are given by

$$
E_j = \hbar \omega \left(j + \frac{1}{2}\right), \qquad j = 0,1,2,\ldots
\tag{1}
$$

The canonical partition function is defined as

$$
Z(T) = \sum_{j=0}^{\infty} e^{-\frac{E_j}{k_B T}}
\tag{2}
$$

Substituting Equation (1) into Equation (2),

$$
Z(T)=\sum_{j=0}^{\infty}
e^{-\frac{\hbar\omega\left(j+\frac12\right)}{k_B T}}
$$

Separate the exponential into two factors:

$$
Z(T)=
\sum_{j=0}^{\infty}
e^{-\frac{\hbar\omega}{2k_B T}}
e^{-\frac{\hbar\omega j}{k_B T}}
$$

Since the first exponential is independent of $j$, it can be taken outside the summation:

$$
Z(T)=
e^{-\frac{\hbar\omega}{2k_B T}}
\sum_{j=0}^{\infty}
\left(e^{-\frac{\hbar\omega}{k_B T}}\right)^j
$$

The remaining summation is an infinite geometric series of the form

$$
\sum_{j=0}^{\infty} r^j = \frac{1}{1-r},
$$

where

$$
r=e^{-\frac{\hbar\omega}{k_B T}}.
$$

Since the absolute temperature satisfies $T>0$, the common ratio obeys

$$
0<r<1,
$$

so the series converges.

Therefore, the canonical partition function becomes

$$
Z(T)=
\frac{e^{-\frac{\hbar\omega}{2k_B T}}}
{1-e^{-\frac{\hbar\omega}{k_B T}}}.
$$

## 2. Helmholtz Free Energy and Internal Energy

The Helmholtz free energy is defined as

$$
F(T) = -k_B T \ln Z(T).
$$

Using the partition function obtained previously,

$$
Z(T)=
\frac{e^{-\frac{\hbar\omega}{2k_B T}}}
{1-e^{-\frac{\hbar\omega}{k_B T}}},
$$

the Helmholtz free energy becomes

$$
F(T)=
-k_B T
\ln\left(
\frac{e^{-\frac{\hbar\omega}{2k_B T}}}
{1-e^{-\frac{\hbar\omega}{k_B T}}}
\right).
$$

Applying the logarithmic identity
$\ln\left(\frac{a}{b}\right)=\ln a-\ln b$,

$$
F(T)=
-k_B T
\left[
\ln\left(e^{-\frac{\hbar\omega}{2k_B T}}\right)
-
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
\right].
$$

Since

$$
\ln\left(e^x\right)=x,
$$

the expression simplifies to

$$
F(T)=
-k_B T
\left[
-\frac{\hbar\omega}{2k_B T}
-
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
\right].
$$

Therefore, the Helmholtz free energy is

$$
F(T)=
\frac{\hbar\omega}{2}
+
k_B T
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right).
$$

---

### Internal Energy

Introduce the thermodynamic parameter

$$
\beta=\frac{1}{k_B T}.
$$

The partition function can then be written as

$$
Z=
\frac{e^{-\beta\hbar\omega/2}}
{1-e^{-\beta\hbar\omega}}.
$$

Taking the natural logarithm gives

$$
\ln Z=
-\frac{\beta\hbar\omega}{2}
-
\ln\left(1-e^{-\beta\hbar\omega}\right).
$$

The internal energy is obtained from

$$
U=
-\frac{\partial \ln Z}{\partial\beta}.
$$

Differentiating with respect to $\beta$,

$$
U=
-\left[
-\frac{\hbar\omega}{2}
-
\frac{1}{1-e^{-\beta\hbar\omega}}
\left(-e^{-\beta\hbar\omega}\right)
(-\hbar\omega)
\right].
$$

Simplifying,

$$
U=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega e^{-\beta\hbar\omega}}
{1-e^{-\beta\hbar\omega}}.
$$

Multiplying both the numerator and denominator of the second term by $e^{\beta\hbar\omega}$,

$$
U=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\beta\hbar\omega}-1}.
$$

Finally, substituting

$$
\beta=\frac{1}{k_B T},
$$

the internal energy as a function of temperature is

$$
U(T)=
\hbar\omega
\left(
\frac{1}{2}
+
\frac{1}
{e^{\frac{\hbar\omega}{k_B T}}-1}
\right).
$$

## 3. Entropy and Verification of the Thermodynamic Relation

The entropy of the system can be obtained from the Helmholtz free energy using

$$
S(T)=-\frac{\partial F}{\partial T}.
$$

Using the previously derived expression for the Helmholtz free energy,

$$
F(T)=
\frac{\hbar\omega}{2}
+
k_B T
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right),
$$

the entropy is given by

$$
S(T)=
-\frac{\partial}{\partial T}
\left[
\frac{\hbar\omega}{2}
+
k_B T
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
\right].
$$

The first term,
$\frac{\hbar\omega}{2}$, is independent of temperature, so its derivative is zero.

For the second term, applying the product rule,

$$
S(T)=
-\left[
k_B
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
+
k_B T
\frac{1}{1-e^{-\frac{\hbar\omega}{k_B T}}}
\left(-e^{-\frac{\hbar\omega}{k_B T}}\right)
\left(\frac{\hbar\omega}{k_B T^2}\right)
\right].
$$

After simplification,

$$
S(T)=
-k_B
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
+
\frac{\hbar\omega}{T}
\frac{e^{-\frac{\hbar\omega}{k_B T}}}
{1-e^{-\frac{\hbar\omega}{k_B T}}}.
$$

Using the identity

$$
\frac{e^{-x}}{1-e^{-x}}=\frac{1}{e^x-1},
$$

the entropy expression becomes

$$
S(T)=
-k_B
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
+
\frac{\hbar\omega}{T}
\frac{1}
{e^{\frac{\hbar\omega}{k_B T}}-1}.
$$

---

## Verification of the Thermodynamic Relation

The thermodynamic relation between internal energy, Helmholtz free energy, and entropy is

$$
U=F+TS.
$$

Multiplying the entropy expression by temperature,

$$
TS=
-k_B T
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
+
\frac{\hbar\omega}
{e^{\frac{\hbar\omega}{k_B T}}-1}.
$$

Now substituting $F(T)$ and $TS$,

$$
F+TS=
\left[
\frac{\hbar\omega}{2}
+
k_B T
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
\right]
+
\left[
-k_B T
\ln\left(1-e^{-\frac{\hbar\omega}{k_B T}}\right)
+
\frac{\hbar\omega}
{e^{\frac{\hbar\omega}{k_B T}}-1}
\right].
$$

The logarithmic terms cancel each other, leaving

$$
F+TS=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\frac{\hbar\omega}{k_B T}}-1}.
$$

Therefore,

$$
F+TS=
\hbar\omega
\left(
\frac{1}{2}
+
\frac{1}
{e^{\frac{\hbar\omega}{k_B T}}-1}
\right),
$$

which is exactly the previously derived expression for the internal energy:

$$
U(T)=
\hbar\omega
\left(
\frac{1}{2}
+
\frac{1}
{e^{\frac{\hbar\omega}{k_B T}}-1}
\right).
$$

Hence, the thermodynamic identity

$$
\boxed{U=F+TS}
$$

is successfully verified.

## 4. Heat Capacity of a Quantum Harmonic Oscillator

The heat capacity at constant volume is defined as

$$
C_V=\left(\frac{\partial U}{\partial T}\right)_V .
$$

Using the expression for internal energy,

$$
U(T)=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\frac{\hbar\omega}{k_B T}}-1},
$$

the heat capacity can be calculated as

$$
C_V=
\frac{\partial}{\partial T}
\left[
\frac{\hbar\omega}{2}
+
\hbar\omega
\left(
e^{\frac{\hbar\omega}{k_B T}}-1
\right)^{-1}
\right].
$$

The first term is independent of temperature, therefore its derivative is zero. Applying the chain rule to the second term,

$$
C_V=
\hbar\omega(-1)
\left(
e^{\frac{\hbar\omega}{k_B T}}-1
\right)^{-2}
e^{\frac{\hbar\omega}{k_B T}}
\left(
-\frac{\hbar\omega}{k_B T^2}
\right).
$$

Simplifying the expression,

$$
C_V=
\frac{(\hbar\omega)^2}{k_B T^2}
\frac{
e^{\frac{\hbar\omega}{k_B T}}
}
{
\left(
e^{\frac{\hbar\omega}{k_B T}}-1
\right)^2
}.
$$

Rearranging in terms of the dimensionless quantity
$\frac{\hbar\omega}{k_B T}$,

$$
\boxed{
C_V=
k_B
\left(
\frac{\hbar\omega}{k_B T}
\right)^2
\frac{
e^{\frac{\hbar\omega}{k_B T}}
}
{
\left(
e^{\frac{\hbar\omega}{k_B T}}-1
\right)^2
}
}
$$

---

## Hyperbolic Function Representation

To express the heat capacity in hyperbolic form, define

$$
x=\frac{\hbar\omega}{2k_B T}.
$$

Therefore,

$$
\frac{\hbar\omega}{k_B T}=2x.
$$

Substituting this into the exponential term,

$$
\frac{e^{2x}}
{(e^{2x}-1)^2}.
$$

Using the identity,

$$
e^{2x}-1=e^x(e^x-e^{-x}),
$$

we obtain

$$
(e^{2x}-1)^2
=
e^{2x}(e^x-e^{-x})^2.
$$

Hence,

$$
\frac{e^{2x}}
{(e^{2x}-1)^2}
=
\frac{e^{2x}}
{e^{2x}(e^x-e^{-x})^2}
=
\frac{1}
{(e^x-e^{-x})^2}.
$$

Using the hyperbolic sine identity,

$$
e^x-e^{-x}=2\sinh(x),
$$

the denominator becomes

$$
(2\sinh(x))^2=4\sinh^2(x).
$$

Therefore,

$$
\frac{e^{2x}}
{(e^{2x}-1)^2}
=
\frac{1}{4\sinh^2(x)}.
$$

Substituting this result into the heat capacity expression,

$$
C_V=
k_B(2x)^2
\left(
\frac{1}{4\sinh^2(x)}
\right).
$$

Simplifying,

$$
C_V=
k_B
\frac{x^2}{\sinh^2(x)}.
$$

Finally, replacing $x$ with its original definition,

$$
\boxed{
C_V=
k_B
\left(
\frac{\hbar\omega}{2k_B T}
\right)^2
\frac{1}
{
\sinh^2
\left(
\frac{\hbar\omega}{2k_B T}
\right)
}
}
$$

This is the hyperbolic form of the heat capacity for a single quantum harmonic oscillator.

## 5. High-Temperature Limit and Classical Behavior

In the high-temperature limit, the thermal energy becomes much larger than the quantum energy spacing:

$$
k_B T \gg \hbar\omega .
$$

Using the dimensionless parameter defined previously,

$$
x=\frac{\hbar\omega}{2k_B T},
$$

this condition implies that

$$
x \rightarrow 0.
$$

For small values of $x$, the Taylor series expansion of the hyperbolic sine function can be approximated as

$$
\sinh(x)\approx x.
$$

Substituting this approximation into the hyperbolic form of the heat capacity,

$$
C_V=
k_B
\frac{x^2}{\sinh^2(x)},
$$

gives

$$
C_V
\approx
k_B
\frac{x^2}{x^2}.
$$

Therefore,

$$
\boxed{C_V \approx k_B}
$$

This result agrees with the prediction of the classical equipartition theorem.

According to the classical equipartition theorem, each independent quadratic degree of freedom contributes

$$
\frac{1}{2}k_B
$$

to the heat capacity of a system.

A one-dimensional harmonic oscillator contains two quadratic energy terms:

- Kinetic energy term
- Potential energy term

Each contributes $\frac{1}{2}k_B$, resulting in

$$
C_V=
2\times\frac{1}{2}k_B
=
k_B.
$$

Therefore, at high temperatures, the effects of quantum energy quantization become negligible. The energy levels appear continuous, and the quantum harmonic oscillator approaches the behavior of a classical harmonic oscillator.

# **Q2) The Maximum Entropy Distribution under Mean and Variance Constraints**

## Maximum Entropy Derivation for a Continuous Probability Distribution

The continuous Shannon entropy for a probability density function $f(x)$, assuming a uniform measure function $m(x)=1$, is defined as

$$
h(f)=
-\int_{-\infty}^{\infty}
f(x)\ln f(x)\,dx .
$$

The objective is to find the probability distribution $f(x)$ that maximizes this entropy while satisfying the required constraints.

The maximization is performed under three integral constraints:

### 1. Normalization Constraint

The total probability of the distribution must be equal to one:

$$
\int_{-\infty}^{\infty} f(x)\,dx=1.
$$

This ensures that $f(x)$ represents a valid probability density function.

---

### 2. Mean Constraint

The expected value of the random variable is fixed at $\mu$:

$$
\int_{-\infty}^{\infty} x f(x)\,dx=\mu.
$$

This constraint specifies the average value of the distribution.

---

### 3. Variance Constraint

The spread of the distribution around the mean is constrained by the variance:

$$
\int_{-\infty}^{\infty}
(x-\mu)^2 f(x)\,dx
=
\sigma^2.
$$

This ensures that the distribution has a specified variance $\sigma^2$.

Therefore, the problem is to maximize the Shannon entropy functional while maintaining the normalization, mean, and variance constraints.

## Maximization of Entropy Using Lagrange Multipliers

To maximize the Shannon entropy while satisfying the normalization, mean, and variance constraints, introduce Lagrange multipliers:

- $\lambda_0$ for the normalization constraint,
- $\lambda_1$ for the mean constraint,
- $\lambda_2$ for the variance constraint.

The constrained entropy functional can be written as

$$
\mathcal{L}[f]
=
-\int_{-\infty}^{\infty}
f(x)\ln f(x)\,dx
-\lambda_0
\left(
\int_{-\infty}^{\infty}f(x)\,dx-1
\right)
-\lambda_1
\left(
\int_{-\infty}^{\infty}x f(x)\,dx-\mu
\right)
-\lambda_2
\left(
\int_{-\infty}^{\infty}(x-\mu)^2 f(x)\,dx-\sigma^2
\right).
$$

The maximum entropy distribution is obtained by finding the stationary point of this functional. Therefore, take the functional derivative of $\mathcal{L}[f]$ with respect to $f(x)$ and set it equal to zero:

$$
\frac{\delta\mathcal{L}}{\delta f(x)}=0.
$$

Evaluating the functional derivative gives

$$
-\ln f(x)-1-\lambda_0-\lambda_1x-\lambda_2(x-\mu)^2=0.
$$

Rearranging the equation,

$$
\ln f(x)
=
-1-\lambda_0-\lambda_1x-\lambda_2(x-\mu)^2.
$$

Taking the exponential of both sides,

$$
f(x)
=
\exp
\left(
-1-\lambda_0-\lambda_1x-\lambda_2(x-\mu)^2
\right).
$$

Hence, the maximum entropy probability distribution has an exponential form determined by the imposed constraints.

## Evaluation of the Lagrange Multiplier $\lambda_1$

Since the variance constraint is defined directly with respect to the mean $\mu$, it is convenient to shift the coordinate system and consider deviations from the mean.

The mean constraint can be rewritten as

$$
\int_{-\infty}^{\infty}(x-\mu)f(x)\,dx=0.
$$

This indicates that the distribution is centered around $\mu$, and therefore the linear term in the exponential form does not contribute. Hence, we can choose

$$
\lambda_1=0.
$$

Substituting this into the maximum entropy distribution,

$$
f(x)=
\exp
\left(
-1-\lambda_0-\lambda_2(x-\mu)^2
\right),
$$

which can be separated into two parts:

$$
f(x)=
e^{-1-\lambda_0}
e^{-\lambda_2(x-\mu)^2}.
$$

Define the normalization constant

$$
Z=e^{1+\lambda_0}.
$$

Therefore, the probability distribution becomes

$$
f(x)=
\frac{1}{Z}
e^{-\lambda_2(x-\mu)^2}.
$$

To determine $Z$, apply the normalization condition:

$$
\int_{-\infty}^{\infty}f(x)\,dx=1.
$$

Substituting the expression for $f(x)$,

$$
\int_{-\infty}^{\infty}
\frac{1}{Z}
e^{-\lambda_2(x-\mu)^2}
dx=1.
$$

Using the substitution

$$
u=x-\mu, \qquad du=dx,
$$

the integral becomes

$$
\frac{1}{Z}
\int_{-\infty}^{\infty}
e^{-\lambda_2u^2}\,du=1.
$$

The Gaussian integral has the standard form

$$
\int_{-\infty}^{\infty}e^{-au^2}\,du
=
\sqrt{\frac{\pi}{a}}.
$$

Taking $a=\lambda_2$,

$$
\frac{1}{Z}
\sqrt{\frac{\pi}{\lambda_2}}
=1.
$$

Therefore, the normalization constant is

$$
\boxed{
Z=\sqrt{\frac{\pi}{\lambda_2}}
}
$$

and the maximum entropy distribution can be normalized accordingly.

## Evaluation of the Lagrange Multiplier $\lambda_2$

To determine the value of $\lambda_2$, substitute the maximum entropy distribution

$$
f(x)=
\frac{1}{Z}
e^{-\lambda_2(x-\mu)^2}
$$

into the variance constraint:

$$
\int_{-\infty}^{\infty}
(x-\mu)^2
\frac{1}{Z}
e^{-\lambda_2(x-\mu)^2}
dx
=
\sigma^2.
$$

Introduce the variable transformation

$$
u=x-\mu, \qquad du=dx,
$$

which gives

$$
\frac{1}{Z}
\int_{-\infty}^{\infty}
u^2 e^{-\lambda_2u^2}\,du
=
\sigma^2.
$$

The integral is a standard Gaussian integral of the form

$$
\int_{-\infty}^{\infty}
u^2e^{-au^2}\,du
=
\frac{1}{2a}
\sqrt{\frac{\pi}{a}}.
$$

By taking

$$
a=\lambda_2,
$$

the variance constraint becomes

$$
\frac{1}{Z}
\left(
\frac{1}{2\lambda_2}
\sqrt{\frac{\pi}{\lambda_2}}
\right)
=
\sigma^2.
$$

From the normalization condition,

$$
Z=
\sqrt{\frac{\pi}{\lambda_2}}.
$$

Substituting this value into the variance equation,

$$
\frac{1}
{\sqrt{\frac{\pi}{\lambda_2}}}
\left(
\frac{1}{2\lambda_2}
\sqrt{\frac{\pi}{\lambda_2}}
\right)
=
\sigma^2.
$$

The common terms cancel, resulting in

$$
\frac{1}{2\lambda_2}
=
\sigma^2.
$$

Therefore,

$$
\boxed{
\lambda_2=
\frac{1}{2\sigma^2}
}
$$

Hence, the second Lagrange multiplier is determined entirely by the variance constraint.

## Final Maximum Entropy Distribution

Using the previously obtained value of the Lagrange multiplier,

$$
\lambda_2=\frac{1}{2\sigma^2},
$$

the normalization constant becomes

$$
Z=
\sqrt{
\frac{\pi}{\lambda_2}
}.
$$

Substituting $\lambda_2$ into this expression,

$$
Z=
\sqrt{
\frac{\pi}{\frac{1}{2\sigma^2}}
},
$$

which simplifies to

$$
Z=
\sqrt{2\pi\sigma^2}.
$$

Now substitute the obtained values of $Z$ and $\lambda_2$ into the maximum entropy probability density function,

$$
f(x)=
\frac{1}{Z}
e^{-\lambda_2(x-\mu)^2}.
$$

This gives

$$
f(x)=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp
\left(
-\frac{(x-\mu)^2}{2\sigma^2}
\right).
$$

Therefore, the probability distribution that maximizes the continuous Shannon entropy under fixed mean and variance constraints is

$$
\boxed{
f(x)=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp
\left(
-\frac{(x-\mu)^2}{2\sigma^2}
\right)
}
$$

This is the Gaussian (normal) distribution, demonstrating that the normal distribution is the maximum entropy distribution for a continuous variable with specified mean and variance.

# **Q3) Multiclass Logistic Regression: Automatic Classification of Emails**

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




```python
import numpy as np
from scipy.optimize import minimize
import plotly.graph_objects as go

# Logistic sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# Maximum Entropy objective function
# Minimizes the difference between model expectations and target constraints
def max_entropy_loss(weights):
    w_finance, w_access, w_tech = weights

    # Calculate expected feature values from the MaxEnt model
    expected_x1 = (
        0.40 * sigmoid(w_finance) +
        0.05 * sigmoid(w_finance + w_access)
    )

    expected_x2 = (
        0.25 * sigmoid(w_access) +
        0.15 * sigmoid(w_access + w_tech) +
        0.05 * sigmoid(w_finance + w_access)
    )

    expected_x3 = (
        0.05 * sigmoid(w_tech) +
        0.15 * sigmoid(w_access + w_tech)
    )

    # Observed constraints obtained from historical data
    constraint_x1 = 0.05
    constraint_x2 = 0.15
    constraint_x3 = 0.16

    # Compute squared error between model and constraints
    loss = (
        (expected_x1 - constraint_x1)**2 +
        (expected_x2 - constraint_x2)**2 +
        (expected_x3 - constraint_x3)**2
    )

    return loss


# Initial values for Lagrange parameters
initial_weights = [0.0, 0.0, 0.0]

# Optimize the MaxEnt objective function
optimization_result = minimize(
    max_entropy_loss,
    initial_weights,
    method='L-BFGS-B'
)

# Extract optimized parameters
w_finance, w_access, w_tech = optimization_result.x

print(
    f"Optimized MaxEnt Weights:\n"
    f" w1 (Financial Feature): {w_finance:.4f}\n"
    f" w2 (Access Feature): {w_access:.4f}\n"
    f" w3 (Technical Feature): {w_tech:.4f}"
)


# Define possible email scenarios
scenarios = [
    'A (Finance Only)',
    'B (Access Only)',
    'C (Technical Only)',
    'D (Access + Technical)',
    'E (Finance + Access)'
]

# Probability of routing towards technical category
technical_probability = [
    sigmoid(w_finance),
    sigmoid(w_access),
    sigmoid(w_tech),
    sigmoid(w_access + w_tech),
    sigmoid(w_finance + w_access)
]

# Probability of routing towards finance category
finance_probability = [
    1 - probability for probability in technical_probability
]


# Create stacked bar visualization
figure = go.Figure()

figure.add_trace(
    go.Bar(
        x=scenarios,
        y=technical_probability,
        name='Technical Route (Y=1)',
        marker_color='#1F77B4'
    )
)

figure.add_trace(
    go.Bar(
        x=scenarios,
        y=finance_probability,
        name='Finance Route (Y=0)',
        marker_color='#FF7F0E'
    )
)

# Customize plot appearance
figure.update_layout(
    title='Maximum Entropy Email Routing Probabilities',
    xaxis_title='Email Feature Scenarios',
    yaxis_title='Probability',
    barmode='stack',
    template='plotly_dark',
    yaxis=dict(tickformat=".0%"),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

figure.show()
```

# **Q4) Asset Pricing and Portfolio Optimization.**


In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



```python
import numpy as np
from scipy.optimize import minimize
from scipy.integrate import quad
import plotly.graph_objects as go


# ----------------------------------------------------
# Target market constraints
# ----------------------------------------------------
desired_mean = 102.0          # E[S]
desired_second_moment = 10600.0   # E[S^2]
desired_log_mean = 4.60       # E[ln(S)]


# ----------------------------------------------------
# Unnormalized Maximum Entropy Probability Density
# p(S) ∝ exp(-λ1S - λ2S² - λ3ln(S))
# ----------------------------------------------------
def maxent_density(S, lambda_1, lambda_2, lambda_3):

    exponent = (
        -lambda_1 * S
        -lambda_2 * S**2
        -lambda_3 * np.log(S)
    )

    return np.exp(exponent)


# ----------------------------------------------------
# Objective function for finding Lagrange multipliers
# ----------------------------------------------------
def maxent_error(lambdas):

    lambda_1, lambda_2, lambda_3 = lambdas

    # Calculate normalization constant
    Z, _ = quad(
        maxent_density,
        1e-5,
        np.inf,
        args=(lambda_1, lambda_2, lambda_3)
    )

    # Prevent invalid solutions
    if Z <= 0 or np.isinf(Z) or np.isnan(Z):
        return 1e9


    # Calculate expected values from the distribution
    calculated_mean, _ = quad(
        lambda S: S * maxent_density(S, lambda_1, lambda_2, lambda_3) / Z,
        1e-5,
        np.inf
    )

    calculated_second_moment, _ = quad(
        lambda S: S**2 * maxent_density(S, lambda_1, lambda_2, lambda_3) / Z,
        1e-5,
        np.inf
    )

    calculated_log_mean, _ = quad(
        lambda S: np.log(S) * maxent_density(S, lambda_1, lambda_2, lambda_3) / Z,
        1e-5,
        np.inf
    )


    # Constraint matching error
    loss = (
        (calculated_mean - desired_mean)**2
        + ((calculated_second_moment - desired_second_moment) / 100)**2
        + (calculated_log_mean - desired_log_mean)**2
    )

    return loss


# ----------------------------------------------------
# Optimize Lagrange multipliers
# ----------------------------------------------------
initial_parameters = [0.01, 0.0001, -0.5]

optimization = minimize(
    maxent_error,
    initial_parameters,
    method='Nelder-Mead'
)


lambda_1_opt, lambda_2_opt, lambda_3_opt = optimization.x


print(
    f"Optimized Lagrange Multipliers:\n"
    f" λ1 = {lambda_1_opt:.6f}\n"
    f" λ2 = {lambda_2_opt:.6f}\n"
    f" λ3 = {lambda_3_opt:.6f}"
)


# ----------------------------------------------------
# Construct normalized PDF
# ----------------------------------------------------
normalization_constant, _ = quad(
    maxent_density,
    1e-5,
    np.inf,
    args=(lambda_1_opt, lambda_2_opt, lambda_3_opt)
)


# Price range for visualization
asset_prices = np.linspace(0.1, 250, 500)


probability_density = [
    maxent_density(
        S,
        lambda_1_opt,
        lambda_2_opt,
        lambda_3_opt
    ) / normalization_constant
    for S in asset_prices
]


# ----------------------------------------------------
# Plot Maximum Entropy Risk-Neutral Distribution
# ----------------------------------------------------
fig = go.Figure()


fig.add_trace(
    go.Scatter(
        x=asset_prices,
        y=probability_density,
        mode='lines',
        name='MaxEnt Risk-Neutral PDF',
        line=dict(
            color='#636EFA',
            width=3
        ),
        fill='tozeroy',
        fillcolor='rgba(99,110,250,0.2)'
    )
)


# Mark expected price
fig.add_vline(
    x=desired_mean,
    line_width=2,
    line_dash="dash",
    line_color="#FFA15A",
    annotation_text="Expected Price ($102)",
    annotation_position="top right"
)


fig.update_layout(
    title="Maximum Entropy Risk-Neutral Asset Price Distribution",
    xaxis_title="Asset Price at Expiry (S_T)",
    yaxis_title="Probability Density p(S_T)",
    template="plotly_dark",
    height=600
)


fig.show()
```
