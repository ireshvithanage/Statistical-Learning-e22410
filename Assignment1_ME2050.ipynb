# Question 1

Let the disk be
$$
\Omega=\{(x,y)\in \mathbb{R}^2 : x^2+y^2\le 10^2\}.
$$

Let $A$ be the event that the chosen point lies within $1$ unit of the boundary.  
Since the boundary is the circle of radius $10$, this means the point must satisfy
$$
9 \le \sqrt{x^2+y^2} \le 10.
$$

Thus,
$$
A=\left\{(x,y)\in \Omega : 9\le \sqrt{x^2+y^2}\le 10\right\},
$$
Where $A$ represents the outer annulus of the disk.

---

## 1 : Uniform distribution over area

In this model, equal areas occupie equal probability. Thus, probability is given by:
$$
P(A)=\frac{\text{Area of }A}{\text{Total area of disk}}.
$$

The total area of the disk is
$$
\pi(10)^2=100\pi.
$$

The area of the annulus $A$ is
$$
\pi(10)^2-\pi(9)^2
=100\pi-81\pi
=19\pi.
$$

Therefore,
$$
P(A)=\frac{19\pi}{100\pi}=\frac{19}{100}.
$$

Thus,
$$
\boxed{P(A)=\frac{19}{100}}.
$$

---

## 2 : Uniform distribution of the radius

In this model:
- The radius $r$ is chosen uniformly from $[0,10]$
- The angle $\theta$ is chosen uniformly from $[0,2\pi)$

Since the angle does not affect the distance from the center, we only consider $r$.

- Total possible range of $r$: $[0,10]$ (length = $10$)
- Favorable range for $A$: $[9,10]$ (length = $1$)

Thus,
$$
P(A)=\frac{10-9}{10-0}=\frac{1}{10}.
$$

Therefore,
$$
\boxed{P(A)=\frac{1}{10}}.
$$

---

## Why the two answers differ

The difference arises because the phrase “chosen at random” does not uniquely determine a probability model.

- In the first model, probability is distributed uniformly over **area**.  
  Since outer regions occupy more area, points near the boundary are more likely:
  $$
  P(A)=\frac{19}{100}.
  $$

- In the second model, probability is distributed uniformly over the **radius**.  
  Each radial interval has equal probability regardless of its area:
  $$
  P(A)=\frac{1}{10}.
  $$

Thus, the same event $A$ has different probabilities because the underlying probability measures are different.

---

## Key Insight

This example demonstrates that probabilities are determined not just by the sample space, but also by the **probability measure** applied. Different ways of defining “random selection” can therefore produce different outcomes.

---

## Physical Interpretation (Dart Example)

If points are produced by randomly throwing darts at a circular board with no bias, the probability of landing in a region is proportional to its **area**.
In this case,
$$
P(A)\approx \frac{19}{100}.
$$

The radius-uniform model, however, follows a different process:

* First, a distance ( r ) is chosen uniformly
* Then, a direction is selected

This approach is an artificial sampling method and does not reflect typical real-world randomness.

---

## Conclusion

- Area-uniform model → physically realistic → $\displaystyle P(A)=\frac{19}{100}$
- Radius-uniform model → mathematical construction → $\displaystyle P(A)=\frac{1}{10}$

Hence, the meaning of “random” must always be clearly specified in probability problems.

## Preliminary Geometric Fact

Let the circle be
$$
x^2 + y^2 = 1.
$$

The side length of the inscribed equilateral triangle is
$$
\sqrt{3}.
$$

A chord is longer than $\sqrt{3}$ **if and only if** its perpendicular distance from the center is less than $1/2$.

### Proof

If a chord is at distance $r \in [0,1]$ from the center, then from a right triangle,
$$
L(r) = 2\sqrt{1 - r^2}.
$$

We require
$$
L(r) > \sqrt{3}
\;\;\Longleftrightarrow\;\;
2\sqrt{1 - r^2} > \sqrt{3}
\;\;\Longleftrightarrow\;\;
r < \frac{1}{2}.
$$

Thus, the event of interest is
$$
A = \{ \text{distance from center} < 1/2 \}.
$$

---

## Model 1: Uniform Endpoints (Angle model)

Choose two endpoints uniformly on the circumference.  
By symmetry, fix one endpoint and let $\theta \in [0,\pi]$ be the subtended angle.

- Probability measure:
$$
\Omega_1 = [0,\pi], \quad P_1(E) = \frac{\lambda(E)}{\pi}.
$$

- Chord length:
$$
L(\theta) = 2\sin(\theta/2).
$$

Condition:
$$
2\sin(\theta/2) > \sqrt{3}
\;\;\Longleftrightarrow\;\;
\theta > \frac{2\pi}{3}.
$$

Hence,
$$
A_1 = \left( \frac{2\pi}{3}, \pi \right].
$$

- Probability:
$$
P_1(A_1) = \frac{\pi - 2\pi/3}{\pi} = \frac{1}{3}.
$$

$$
\boxed{P_1 = \frac{1}{3}}
$$

---

## Model 2: Uniform Distance from Center

Choose a radius and then choose a point uniformly along it.  
This makes the perpendicular distance $r$ uniform in $[0,1]$.

- Probability space:
$$
\Omega_2 = [0,1], \quad P_2(E) = \lambda(E).
$$

- Event:
$$
A_2 = [0, 1/2).
$$

- Probability:
$$
P_2(A_2) = \frac{1}{2}.
$$

$$
\boxed{P_2 = \frac{1}{2}}
$$

---

## Model 3: Uniform midpoint in the disk

Choose the midpoint of the chord uniformly in the unit disk.

- Probability space:
$$
\Omega_3 = \{(x,y): x^2 + y^2 \le 1\},
\quad P_3(E) = \frac{\lambda_2(E)}{\pi}.
$$

- Event:
$$
A_3 = \{(x,y): x^2 + y^2 < 1/4\}.
$$

- Probability:
$$
P_3(A_3) = \frac{\pi(1/2)^2}{\pi} = \frac{1}{4}.
$$

$$
\boxed{P_3 = \frac{1}{4}}
$$

---

## Final Comparison

| Model | Random choice                  | Probability |
|------|-------------------------------|------------|
| 1    | Angle / endpoints             | $\frac{1}{3}$ |
| 2    | Distance from center          | $\frac{1}{2}$ |
| 3    | Midpoint in the disk          | $\frac{1}{4}$ |

---

## Conclusion (Bertrand’s Paradox)

Each method produces a different result because the phrase **“choose a chord at random” is inherently ambiguous**.

Different ways of sampling correspond to different probability measures, which in turn lead to different outcomes.



# Question 8: Two-Children Problem

The sample space is
$$
\Omega=\{BB, BG, GB, GG\},
$$
where the first letter represents the older child and the second the younger.

Because each child is equally likely to be a boy (B) or a girl (G), and the outcomes are independent, every possible combination occurs with the same probability.
$$
\frac{1}{4}.
$$

---

## Given Condition

We are given that at least one child is a girl, which rules out only the outcome BB.

Therefore, the sample space is reduced to:
$$
\{BG, GB, GG\}.
$$

Each of these outcomes is still equally likely.

---

## Required Probability

We are interested in the probability that **both children are girls**, meaning the outcome ( GG ).

Among the three possible outcomes, only one meets this condition. Therefore,
$$
P(\text{both girls} \mid \text{at least one girl}) = \frac{1}{3}.
$$

Hence,
$$
\boxed{\frac{1}{3}}.
$$

---

## Why this result seem interesting

A common intuitive answer is ( \frac{1}{2} ), but that result is based on a **different assumption**.

For instance, if we are given the information:

> “The older child is a girl,”

then the set of possible outcomes is reduced to
$$
\{GG, GB\}.
$$

In this case,
$$
P(\text{both girls} \mid \text{older child is a girl}) = \frac{1}{2}.
$$

# Question 10

---

## **1. Probability Space**

The probability space is given by,

$$
\Omega=\mathbb{R}^2, \qquad \mathfrak{F}=\mathfrak{B}(\mathbb{R}^2),
$$

where $\mathfrak{B}(\mathbb{R}^2)$ is the Borel $\sigma$-algebra.

Each outcome is written as

$$
\omega=(t,p),
$$

where:
- $t$ = temperature deviation  
- $p$ = pressure deviation  

The probability measure $P$ is defined through the joint density

$$
f_{T,P}(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right).
$$

Thus, for any Borel set $A \subseteq \mathbb{R}^2$,

$$
P(A)=\iint_A f_{T,P}(t,p)\,dt\,dp.
$$

---

## **2. $\sigma$-algebra generated by $X$**

The random variable

$$
X(t,p)=t
$$

extracts only the temperature component.

The generated $\sigma$-algebra is

$$
\sigma(X)=\{X^{-1}(B): B\in\mathfrak{B}(\mathbb{R})\}.
$$

Since

$$
X^{-1}(B)=B\times \mathbb{R},
$$

we obtain

$$
\sigma(X)=\{B\times \mathbb{R}: B\in\mathfrak{B}(\mathbb{R})\}.
$$

So $\sigma(X)$ contains only vertical-strip information.

---

## **3. $\sigma$-algebra generated by $Y$**

The random vector is

$$
Y(t,p)=(t,p).
$$

Since this is the identity map on $\mathbb{R}^2$,

$$
\sigma(Y)=\mathfrak{B}(\mathbb{R}^2).
$$

---

## **4. Relationship between $\sigma(X)$ and $\sigma(Y)$**

It is evident that
$$
\sigma(X)\subset \sigma(Y).
$$

This inclusion is strict, since sets like
$$
[-1,1]\times[-2,2]
$$
depend on both variables and cannot be described using only ( X ).

Therefore, ( Y ) carries strictly more information than ( X ).


---

## **5. Marginal density of $X$**

$$
f_X(x)=\int_{-\infty}^{\infty} f_{T,P}(x,p)\,dp
$$

which gives

$$
f_X(x)=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}.
$$

So,

$$
X\sim N(0,1).
$$

---

## **6. Marginal density of $Y$**

Since $Y=(T,P)$,

$$
f_Y(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right).
$$

---

## **7. Probability computations**

- $P(X\le 0)=\frac{1}{2}$  
- $P(X>1)=1-\Phi(1)\approx 0.1587$  
- $P(Y\in(-\infty,0]\times\mathbb{R})=\frac{1}{2}$  

Since $T$ and $P$ are independent,

$$
P(Y\in[-1,1]\times[-2,2])=(2\Phi(1)-1)^2\approx 0.466.
$$

---

## **8. Conditional distribution $X|Y$**

Given $Y=(t,p)$,

$$
P(X=t \mid Y=(t,p))=1.
$$

So,

$$
X|Y=\delta_t.
$$

---

## **9. Conditional distribution $Y|X$**

Given $X=t$,

$$
Y|X=t=(t,P), \quad P\sim N(0,4).
$$

Hence the pressure remains random while temperature is fixed.

---

## **10. Interpretation**

* (X) provides only temperature information
* (Y) provides both temperature and pressure
* Hence, (Y) is strictly more informative than (X)
$$
[-1,1]\times[-2,2]
$$
---

## **11. Functional Relationship**

Let

$$
\pi_1(t,p)=t.
$$

Then

$$
X=\pi_1\circ Y.
$$

Hence $X$ is measurable with respect to $Y$, implying

$$
\sigma(X)\subseteq\sigma(Y).
$$

### Python Visulaization

#### The $\sigma$-algebra visualizations

Here is a good Python visualization of the two $\sigma$-algebras in $\mathbb R^2$ for the example

$$
X(t,p)=t, \qquad Y(t,p)=(t,p).
$$

The idea is:

* $\sigma(X)$ consists of sets of the form
  $$
  B\times \mathbb R,
  $$
  so visually these are **vertical strips** in the plane;

* $\sigma(Y)=\mathfrak B(\mathbb R^2)$ consists of all Borel sets in the plane, so visually we can show a general **rectangle**, **disk**, or other 2D region.

This code creates a figure with:

1. a vertical strip representing an event in $\sigma(X)$,
2. a rectangle representing an event in $\sigma(Y)$,
3. an overlay comparison showing that $\sigma(X)$ is coarser than $\sigma(Y)$.




```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle

fig, axes = plt.subplots(1, 3, figsize=(15, 4.8))

xlim = (-4, 4)
ylim = (-8, 8)

# -------------------------
# (1) Event in sigma(X)
# -------------------------
ax = axes[0]
ax.set_title(r"Event in $\sigma(X)$: $B\times\mathbb{R}$")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

# Example B = [-1,1]
strip = Rectangle((-1, ylim[0]), 2, ylim[1] - ylim[0],
                  facecolor="skyblue", edgecolor="navy", alpha=0.55)
ax.add_patch(strip)

ax.text(0, 0, r"$[-1,1]\times \mathbb{R}$",
        ha="center", va="center", fontsize=12,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.grid(alpha=0.25)

# -------------------------
# (2) Event in sigma(Y)
# -------------------------
ax = axes[1]
ax.set_title(r"Event in $\sigma(Y)$: general Borel subset")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

# Example rectangle in sigma(Y)
rect = Rectangle((-1.5, -3), 3, 6,
                 facecolor="lightgreen", edgecolor="darkgreen", alpha=0.55)
ax.add_patch(rect)

# Also add a disk to emphasize generality
disk = Circle((2.2, 3.5), 1.2,
              facecolor="salmon", edgecolor="darkred", alpha=0.45)
ax.add_patch(disk)

ax.text(0, 0, r"$[-1.5,1.5]\times[-3,3]$",
        ha="center", va="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))
ax.text(2.2, 3.5, r"disk",
        ha="center", va="center", fontsize=10,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.grid(alpha=0.25)

# -------------------------
# (3) Comparison
# -------------------------
ax = axes[2]
ax.set_title(r"Comparison: $\sigma(X)\subseteq\sigma(Y)$")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

# Vertical strip for sigma(X)
strip2 = Rectangle((-1, ylim[0]), 2, ylim[1] - ylim[0],
                   facecolor="skyblue", edgecolor="navy", alpha=0.35,
                   label=r"$\sigma(X)$ event")
ax.add_patch(strip2)

# Rectangle for sigma(Y)
rect2 = Rectangle((-1.5, -3), 3, 6,
                  facecolor="lightgreen", edgecolor="darkgreen", alpha=0.5,
                  label=r"$\sigma(Y)$ event")
ax.add_patch(rect2)

ax.text(0, 6.5, r"$\sigma(X)$ sees only vertical strips",
        ha="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))
ax.text(0, -6.5, r"$\sigma(Y)$ sees full 2D regions",
        ha="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.legend(loc="upper right")
ax.grid(alpha=0.25)

plt.tight_layout()
plt.show()
```

> In this model, $\sigma(X)$ contains exactly the events that depend only on the temperature coordinate $t$, so they appear as vertical strips $B\times \mathbb R$. By contrast, $\sigma(Y)=\mathfrak B(\mathbb R^2)$ contains all Borel subsets of the plane, such as rectangles, disks, and more complicated regions. Thus $\sigma(X)$ is strictly coarser than $\sigma(Y)$.


#### The distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal

# Parameters
mu_T, sigma_T = 0.0, 1.0
mu_P, sigma_P = 0.0, 2.0   # since Var(P)=4

# Grids
t = np.linspace(-4, 4, 400)
p = np.linspace(-8, 8, 400)
T, P = np.meshgrid(t, p)

# Joint density of Y = (T, P)
joint_pdf = (
    norm.pdf(T, loc=mu_T, scale=sigma_T) *
    norm.pdf(P, loc=mu_P, scale=sigma_P)
)

# Marginal density of X = T
fx = norm.pdf(t, loc=mu_T, scale=sigma_T)

# Conditional density of P | X=t0
# Independence implies P | X=t0 ~ N(0,4)
t0 = 1.0
cond_p_given_x = norm.pdf(p, loc=mu_P, scale=sigma_P)

# Conditional distribution of X | Y=(t1,p1):
# Degenerate at t1, approximated here by a narrow Gaussian just for visualization
t1, p1 = 1.0, 2.0
eps = 0.08
cond_x_given_y_vis = norm.pdf(t, loc=t1, scale=eps)

# Figure with subplot layout:
# 1
# 3 4
# 5 6
fig = plt.figure(figsize=(12, 12))
gs = fig.add_gridspec(3, 2, height_ratios=[1.1, 1, 1], hspace=0.45, wspace=0.28)

# (1) Joint contour spanning full top row
ax1 = fig.add_subplot(gs[0, :])
cs = ax1.contourf(T, P, joint_pdf, levels=25, cmap="viridis")
ax1.contour(T, P, joint_pdf, levels=10, colors="white", linewidths=0.5, alpha=0.6)
ax1.set_title(r"(1) Joint density of $Y=(T,P)$")
ax1.set_xlabel(r"$t$")
ax1.set_ylabel(r"$p$")
fig.colorbar(cs, ax=ax1, shrink=0.9, label="Density")

# (3) Marginal of X
ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(t, fx, color="navy", lw=2)
ax3.fill_between(t, fx, color="skyblue", alpha=0.35)
ax3.set_title(r"(3) Marginal density of $X=T$")
ax3.set_xlabel(r"$x$")
ax3.set_ylabel(r"$f_X(x)$")
ax3.grid(alpha=0.25)

# (4) Marginal of P
ax4 = fig.add_subplot(gs[1, 1])
fp = norm.pdf(p, loc=mu_P, scale=sigma_P)
ax4.plot(p, fp, color="darkgreen", lw=2)
ax4.fill_between(p, fp, color="lightgreen", alpha=0.35)
ax4.set_title(r"(4) Marginal density of $P$")
ax4.set_xlabel(r"$p$")
ax4.set_ylabel(r"$f_P(p)$")
ax4.grid(alpha=0.25)

# (5) Conditional density of P | X=t0
ax5 = fig.add_subplot(gs[2, 0])
ax5.plot(p, cond_p_given_x, color="crimson", lw=2,
         label=rf"$f_{{P\mid X={t0}}}(p)$")
ax5.fill_between(p, cond_p_given_x, color="salmon", alpha=0.35)
ax5.set_title(r"(5) Conditional density of $P \mid X=t_0$")
ax5.set_xlabel(r"$p$")
ax5.set_ylabel(r"$f_{P\mid X=t_0}(p)$")
ax5.legend()
ax5.grid(alpha=0.25)

# (6) Visualized conditional of X | Y=(t1,p1)
ax6 = fig.add_subplot(gs[2, 1])
ax6.plot(t, cond_x_given_y_vis, color="purple", lw=2,
         label=rf"visualization of $X\mid Y=({t1},{p1})$")
ax6.axvline(t1, color="black", ls="--", lw=1.3, label=rf"$x={t1}$")
ax6.set_title(r"(6) $X\mid Y=(t_1,p_1)$ is degenerate at $x=t_1$")
ax6.set_xlabel(r"$x$")
ax6.set_ylabel("Approx. density")
ax6.legend()
ax6.grid(alpha=0.25)

plt.show()
```

A small note for subplot (6): in this model, $X \mid Y=(t_1,p_1)$ is not an ordinary density but a point mass at $x=t_1$. The code uses a very narrow Gaussian only to make that visible on a plot.
