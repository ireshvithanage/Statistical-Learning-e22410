### Q1.3

By definition, $E[X \mid Y]$ must be $\sigma(Y)$-measurable, meaning it can depend only on the observed value of $Y$ and not on the underlying outcome $\omega$. In other words, there exists a measurable function $g$ such that $E[X \mid Y] = g(Y)$.

Since the $\sigma$-algebra generated by $Y$ can only distinguish events based on the value of $Y$, it is limited to separating outcomes according to the information contained in $Y$ alone.

### Q1.8

We compute the Minimum Mean Square Error (MMSE), which represents the optimal error when using the estimate $\mathbb{E}[X|Y]$.

$$
\text{MSE}_{\text{min}}
=
\mathbb{E}[(X-\mathbb{E}[X\mid Y])^2]
=
\mathbb{E}[\mathrm{Var}(X\mid Y)].
$$

First, we find the conditional variance for each risk class.

For $Y=0$,

$$
\mathrm{Var}(X\mid Y=0)
=
\frac{1}{6}\left(1-\frac{1}{6}\right)
=
\frac{1}{6}\cdot\frac{5}{6}
=
\frac{5}{36}.
$$

For $Y=1$,

$$
\mathrm{Var}(X\mid Y=1)
=
\frac{1}{2}\left(1-\frac{1}{2}\right)
=
\frac{1}{2}\cdot\frac{1}{2}
=
\frac{1}{4}.
$$

Now take the expectation over the distribution of $Y$:

$$
\begin{aligned}
\mathbb{E}[\mathrm{Var}(X\mid Y)]
&=
\mathrm{Var}(X\mid Y=0)P(Y=0)
+
\mathrm{Var}(X\mid Y=1)P(Y=1)
\\
&=
\left(\frac{5}{36}\cdot 0.60\right)
+
\left(\frac{1}{4}\cdot 0.40\right)
\\
&=
\left(\frac{5}{36}\cdot\frac{3}{5}\right)
+
\left(\frac{1}{4}\cdot\frac{2}{5}\right)
\\
&=
\frac{1}{12}+\frac{1}{10}
=
\frac{5+6}{60}
=
\frac{11}{60}
\approx0.1833.
\end{aligned}
$$

Thus,

$$
\boxed{
\text{MSE}_{\text{min}}
=
\frac{11}{60}
\approx0.1833
}
$$

This value is the lowest possible average squared prediction error.

### Q1.10 Quantifying Information Gain

To calculate the information-theoretic quantities, first summarize the joint distribution.

---

### 1. Joint Distribution $P(X,Y)$

Based on the mappings $X(\omega)$ and $Y(\omega)$,

$$
P(X=0,Y=0)=P(\{\omega_1\})=0.50
$$

$$
P(X=1,Y=0)=P(\{\omega_2\})=0.10
$$

$$
P(X=0,Y=1)=P(\{\omega_3\})=0.20
$$

$$
P(X=1,Y=1)=P(\{\omega_4\})=0.20
$$

The marginal distributions are

$$
P(Y=0)=0.60,
\qquad
P(Y=1)=0.40
$$

and

$$
P(X=0)=0.70,
\qquad
P(X=1)=0.30.
$$

---

### 2. Conditional Entropy $H(X\mid Y)$

Conditional entropy is the weighted average of the entropy within each group.

### Step A: Entropy of $X$ given $Y=0$

$$
P(X=1\mid Y=0)=\frac{1}{6},
\qquad
P(X=0\mid Y=0)=\frac{5}{6}
$$

$$
H(X\mid Y=0)
=
-\left[
\frac{1}{6}\log_2\frac{1}{6}
+
\frac{5}{6}\log_2\frac{5}{6}
\right]
\approx0.650\text{ bits}
$$

### Step B: Entropy of $X$ given $Y=1$

$$
P(X=1\mid Y=1)=\frac{1}{2},
\qquad
P(X=0\mid Y=1)=\frac{1}{2}
$$

$$
H(X\mid Y=1)
=
-\left[
\frac{1}{2}\log_2\frac{1}{2}
+
\frac{1}{2}\log_2\frac{1}{2}
\right]
=
1.00\text{ bit}
$$

### Step C: Final Conditional Entropy

$$
H(X\mid Y)
=
P(Y=0)H(X\mid Y=0)
+
P(Y=1)H(X\mid Y=1)
$$

$$
H(X\mid Y)
=
(0.60\times0.650)
+
(0.40\times1.00)
=
0.390+0.40
=
0.790\text{ bits}
$$

---

### 3. Mutual Information $I(X;Y)$

Mutual Information is the reduction in uncertainty:

$$
I(X;Y)
=
H(X)-H(X\mid Y).
$$

### Step A: Calculate $H(X)$

$$
H(X)
=
-\left(
0.70\log_2 0.70
+
0.30\log_2 0.30
\right)
\approx0.881\text{ bits}
$$

### Step B: Subtract $H(X\mid Y)$

$$
I(X;Y)
=
0.881-0.790
=
0.091\text{ bits}
$$

---

### 4. Interpretation

* $H(X)=0.881$ bits is the total uncertainty before observing the risk class.

* $H(X\mid Y)=0.790$ bits is the uncertainty remaining after observing the risk class.

* $I(X;Y)=0.091$ bits is the information gain.

The percentage reduction in uncertainty is

$$
\frac{0.091}{0.881}\times100\%
\approx10.3\%.
$$

Thus, knowing the customer's risk class reduces the uncertainty about fraud by approximately $10.3\%$.

# Q2) Conditional Density of the Multivariate Gaussian

To derive the conditional density $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we start from the definition of conditional probability for continuous random variables:

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) =
\frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}
$$

Using the previously obtained expressions for the joint density $f_{\mathbf{X}, \mathbf{Y}}$ and the marginal density $f_{\mathbf{Y}}$, the derivation reduces to simplifying the ratio. In particular, most of the work comes from combining and rearranging quadratic terms in the exponent.

---

From our previous derivation, we factored the joint quadratic form $Q$ using the Schur complement of $\boldsymbol{\Sigma}_{YY}$:
$$Q_{joint} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) + (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

Where:
* $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

The marginal density $f_{\mathbf{Y}}(\mathbf{y})$ has the exponent:
$$Q_{marginal} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y)$$

When we divide the densities, we subtract the exponents ($e^{A} / e^{B} = e^{A-B}$):
$$Q_{conditional} = Q_{joint} - Q_{marginal} = (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

---

Now we look at the coefficients.
* **Joint constant:** $\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}$
* **Marginal constant:** $\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}$

Dividing the joint constant by the marginal constant gives:
$$\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \cdot \sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|} = \frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}$$

Using the determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the term $\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}$ simplifies exactly to $|\mathbf{S}_{XX}|$.

---

Combining the simplified constant and the remaining quadratic form, we get:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

This is the density of a multivariate normal distribution:
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$

---

**Summary of the Parameters**
* **Conditional Mean:** $\boldsymbol{\mu}_{X|Y} = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$
    * *Insight:* The mean is adjusted by how much $\mathbf{y}$ deviates from its expected value, scaled by the "regression" coefficient $\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$.
* **Conditional Covariance:** $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
    * *Insight:* Notice this does **not** depend on the specific value of $\mathbf{y}$. The variance is always reduced (or stays the same) compared to the marginal $\boldsymbol{\Sigma}_{XX}$.

# Q3) Continuous Fault Monitoring of a Spindle Motor

### 1. Joint Distribution of $(X, Y)$

Since $X$ and $Q$ are independent normal random variables, their sum $Y = X + Q$ is also normally distributed.

- **Mean:**  
  $E[Y] = E[X] + E[Q] = 0 + 0 = 0$

- **Variance:**  
  Since $X \perp Q$,  
  $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$

- **Covariance:**  
  $\text{Cov}(X, Y) = E[XY] - E[X]E[Y] = E[X(X+Q)] = E[X^2] + E[XQ]$

  - $E[X^2] = \text{Var}(X) = 1$
  - $E[XQ] = E[X]E[Q] = 0$

  So,  
  $\textbf{Cov}(X, Y) = 1$

The joint distribution is:

$$
\begin{pmatrix}
X \\
Y
\end{pmatrix}
\sim
\mathscr{N}
\left(
\begin{pmatrix}
0 \\
0
\end{pmatrix},
\begin{pmatrix}
1 & 1 \\
1 & 5
\end{pmatrix}
\right)
$$

---

### 2. Information Geometry

$X$ is not $\sigma(Y)$-measurable because there is no deterministic function $g$ that can uniquely determine $X$ from $Y$, i.e., $X \neq g(Y)$. For any fixed observation $Y = y$, there exist infinitely many combinations of $(x, q)$ satisfying $x + q = y$, so the value of $X$ cannot be uniquely recovered from $Y$ alone.

**Practically:**
A large power reading ($Y$) is inherently ambiguous. It may result from a high internal temperature ($X$), indicating a thermal fault, or from a large mechanical load ($Q$), such as cutting very hard material. Since both effects contribute to the same observation, the sensor reading $Y$ alone cannot distinguish between these underlying causes.

---

### 3. Conditional Expectation (The Best Estimate)

For bivariate normal variables:

$$
E[X \mid Y = y] =
\mu_X +
\frac{\text{Cov}(X, Y)}{\text{Var}(Y)}(y - \mu_Y)
= 0 + \frac{1}{5}(y - 0)
$$

So,

$$
\mathbf{E[X \mid Y] = 0.2Y}
$$

**Interpretation:**  
Only 20% of the variation in $Y$ is attributed to $X$ (temperature).

---

### 4. Verification of the Residual

Let $\widehat{X} = 0.2Y$, and define $Z = X - 0.2Y$.

- **Mean:**
  $E[Z] = E[X] - 0.2E[Y] = 0$

- **Orthogonality:**
  $$
  \text{Cov}(Z, Y) = E[(X - 0.2Y)Y]
  = E[XY] - 0.2E[Y^2]
  $$

  - $E[XY] = 1$
  - $E[Y^2] = 5$

  So,
  $\text{Cov}(Z, Y) = 1 - 1 = 0$

The error is uncorrelated with the observation.

---

### 5. Differential Entropy

Formula:
$$
h(\mathcal{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2)
$$

- **Entropy of $X$:**
  $$
  h(X) = \frac{1}{2}\ln(2\pi e)
  \approx 1.4189 \text{ nats}
  $$

- **Conditional Entropy:**
  $$
  \sigma_{X|Y}^2 = 1 - \frac{1}{5} = 0.8
  $$

  $$
  h(X|Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8)
  \approx 1.3073 \text{ nats}
  $$

---

### 6. Mutual Information

$$
I(X;Y) = h(X) - h(X|Y)
= \frac{1}{2}\ln\left(\frac{1}{0.8}\right)
\approx 0.1116 \text{ nats}
$$

- About **20% variance reduction**
- If noise increases to $\sigma_Q^2 = 100$, then $I(X;Y) \to 0$

---

### 7. Diagnostic Interpretation

- Naive estimator: $X = Y$
- Error:
  $$
  E[(X-Y)^2] = E[Q^2] = 4
  $$

- Optimal estimator:
  $$
  E[X|Y] = 0.2Y
  $$

- Error variance:
  $0.8$

**Conclusion:**  
The optimal estimator “shrinks” noisy measurements toward the mean to reduce error.

## Simulation Insight

The model shows how a noisy sensor $Y = X + Q$ can still produce a strong estimate of $X$ through conditional expectation, even when noise dominates.

````python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal, norm

# --- 1. Parameters for the Spindle Problem ---
# X: Temperature ~ N(0, 1)
# Q: Torque ~ N(0, 4)
# Y: Power = X + Q => Y ~ N(0, 5)
mu_x, mu_y = 0, 0
var_x, var_q = 1, 4
var_y = var_x + var_q
cov_xy = var_x  # Cov(X, X+Q) = Var(X) = 1
rho = cov_xy / (np.sqrt(var_x) * np.sqrt(var_y))

# Shrinkage Factor (Gain)
k = cov_xy / var_y  # 1/5 = 0.2

# --- 2. Generate Data for Visualization ---
# Sample points to show the "Cloud" of data
np.random.seed(42)
n_samples = 1000
x_samples = np.random.normal(mu_x, np.sqrt(var_x), n_samples)
q_samples = np.random.normal(0, np.sqrt(var_q), n_samples)
y_samples = x_samples + q_samples
x_hat = k * y_samples  # E[X|Y]
residual = x_samples - x_hat

# Grids for Contours
x_grid = np.linspace(-4, 4, 200)
y_grid = np.linspace(-8, 8, 200)
X_m, Y_m = np.meshgrid(x_grid, y_grid)
pos = np.dstack((X_m, Y_m))
Z = multivariate_normal([mu_x, mu_y], [[var_x, cov_xy], [cov_xy, var_y]]).pdf(pos)

# --- 3. Create Simulation Plot ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Joint Distribution: X vs YThe Line of Best Estimate (Shrinkage)",
        "Estimation Error (Residual) vs YVisualizing the Orthogonality Principle"
    )
)

# --- LEFT PLOT: Joint X, Y and the Estimator ---
# Density Contour
fig.add_trace(go.Contour(
    x=x_grid, y=y_grid, z=Z, colorscale='Blues', showscale=False, opacity=0.6,
    hoverinfo='skip'
), row=1, col=1)

# Scatter of actual realizations
fig.add_trace(go.Scatter(
    x=x_samples, y=y_samples, mode='markers',
    marker=dict(color='rgba(0,0,0,0.1)', size=3),
    name="Realized (x, y)"
), row=1, col=1)

# The Best Estimate Line E[X|Y] = 0.2Y
y_line = np.linspace(-8, 8, 10)
fig.add_trace(go.Scatter(
    x=k * y_line, y=y_line,
    line=dict(color='Red', width=4),
    name="E[X|Y] = 0.2Y (Optimal)"
), row=1, col=1)

# The Naive Estimator Line X = Y
fig.add_trace(go.Scatter(
    x=y_line, y=y_line,
    line=dict(color='Black', width=2, dash='dash'),
    name="X = Y (Naive)"
), row=1, col=1)

# --- RIGHT PLOT: Orthogonality (Residual vs Y) ---
fig.add_trace(go.Scatter(
    x=y_samples, y=residual, mode='markers',
    marker=dict(color='Purple', opacity=0.4, size=4),
    name="Residual (X - 0.2Y)"
), row=1, col=2)

# Horizontal line at 0 to show no correlation
fig.add_trace(go.Scatter(
    x=y_line, y=np.zeros_like(y_line),
    line=dict(color='Black', width=2),
    name="Zero Mean Error"
), row=1, col=2)

# --- Layout and Styling ---
fig.update_layout(
    height=600,
    title_text="Spindle Motor Diagnostic Simulation: Extracting Signal from Noise",
    template="plotly_white",
    hovermode='x'
)

fig.update_xaxes(title_text="True Temperature (X)", row=1, col=1)
fig.update_yaxes(title_text="Observed Power (Y)", row=1, col=1)
fig.update_xaxes(title_text="Observed Power (Y)", row=1, col=2)
fig.update_yaxes(title_text="Residual Error (Z)", row=1, col=2)

fig.show()
````
