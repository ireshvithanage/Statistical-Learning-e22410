# Part A : Maximum Likelihood & Decision Space (Maximum Likelihood & Decision Space)

## 1. Maximum Likelihood Estimator & Bessel's Correction

Let the true (unknown) population mean be denoted by $\mu$. The covariance Maximum Likelihood Estimator (MLE) is computed using the sample mean $\hat{\mu}_n$. To examine its bias, first express the deviation of each observation from the sample mean as

$$
X_i - \hat{\mu}_n = (X_i - \mu) - (\hat{\mu}_n - \mu).
$$

Substituting this expression into the covariance MLE gives

$$
\hat{\Sigma}_{MLE}
=
\frac{1}{n}
\sum_{i=1}^{n}
\Big[(X_i-\mu)-(\hat{\mu}_n-\mu)\Big]
\Big[(X_i-\mu)-(\hat{\mu}_n-\mu)\Big]^T.
$$

Since

$$
\sum_{i=1}^{n}(X_i-\mu)
=
n(\hat{\mu}_n-\mu),
$$

the expression simplifies to

$$
\hat{\Sigma}_{MLE}
=
\frac{1}{n}
\left[
\sum_{i=1}^{n}(X_i-\mu)(X_i-\mu)^T
-
n(\hat{\mu}_n-\mu)(\hat{\mu}_n-\mu)^T
\right].
$$

Taking expectations, we use the facts that

$$
E[(X_i-\mu)(X_i-\mu)^T] = \Sigma,
$$

and

$$
E[(\hat{\mu}_n-\mu)(\hat{\mu}_n-\mu)^T]
=
\frac{1}{n}\Sigma,
$$

where $\Sigma$ is the true population covariance matrix.

Substituting these results yields

$$
E[\hat{\Sigma}_{MLE}]
=
\frac{1}{n}
\left(
n\Sigma
-
n\left(\frac{1}{n}\Sigma\right)
\right)
=
\frac{1}{n}(n\Sigma-\Sigma)
=
\frac{n-1}{n}\Sigma.
$$

Thus,

$$
E[\hat{\Sigma}_{MLE}]
=
\frac{n-1}{n}\Sigma
\neq
\Sigma.
$$

This demonstrates that the covariance MLE systematically underestimates the true covariance matrix. The bias arises because the deviations are measured relative to the sample mean, which is estimated from the same data and therefore reduces the observed variability.

## Applying Bessel's Correction

To remove this bias, the estimator is multiplied by the correction factor

$$
\frac{n}{n-1},
$$

commonly known as **Bessel's correction**. The unbiased covariance estimator becomes

$$
S
=
\frac{n}{n-1}\hat{\Sigma}_{MLE}
=
\frac{1}{n-1}
\sum_{i=1}^{n}
(X_i-\hat{\mu}_n)(X_i-\hat{\mu}_n)^T.
$$

Taking expectations,

$$
E[S]
=
\frac{n}{n-1}
E[\hat{\Sigma}_{MLE}]
=
\frac{n}{n-1}
\left(
\frac{n-1}{n}\Sigma
\right)
=
\Sigma.
$$

Therefore, Bessel's correction exactly compensates for the downward bias of the MLE covariance estimator, making $S$ an unbiased estimator of the true population covariance matrix.

## 2. Diagnostic Decision Boundaries and Operating Thresholds



## 2. Diagnostic Decision Boundaries and Operating Thresholds

### Defining Diagnostic Errors

* **Type I Error ($\alpha$): False Alarm**  
  A Type I error occurs when the system incorrectly rejects the null hypothesis. In this case, the diagnostic model indicates that a physical asset is damaged or failing, even though it is actually operating under normal conditions.

* **Type II Error ($\beta$): Missed Detection**  
  A Type II error occurs when the system fails to reject the null hypothesis even though a true fault, such as a structural crack or anomaly, is present. This results in a failure to detect a potentially dangerous condition in the asset.

---

### Mathematical Consequences of an Extremely Low Threshold ($\alpha = 0.0001$)

If the significance level is reduced to $\alpha = 0.0001$ in an attempt to eliminate false alarms and avoid unnecessary shutdowns, the decision boundary shifts significantly into the extreme tail regions of the baseline distribution.

### Impact on Statistical Power

Setting such a strict rejection criterion for $H_0$ substantially increases the probability of a Type II error ($\beta$). As $\beta$ increases, the statistical power of the diagnostic system ($1 - \beta$) decreases sharply. This weakens the system’s ability to detect early-stage structural deterioration, effectively reducing its sensitivity to real faults.

### Impact on the Confidence Ellipsoid

The geometric region defining "normal operation" expands to correspond with the confidence level $1 - \alpha = 0.9999$. As a result, the confidence ellipsoid grows significantly in size, absorbing a much larger portion of natural system variability. This makes the diagnostic framework less responsive to meaningful deviations, allowing potentially critical changes in system behavior to be classified as normal.

# Part B : Theoretical Extension (Asymptotic Distribution & Slutsky's Theorem)

## 1. Slutsky’s Theorem

Let a sequence of random variables (or vectors) satisfy:

- $X_n \xrightarrow{d} X$ (convergence in distribution)  
- $Y_n \xrightarrow{p} c$ (convergence in probability to a constant matrix or scalar)

Then standard asymptotic operations preserve convergence in distribution:

* $X_n + Y_n \xrightarrow{d} X + c$
* $X_n Y_n \xrightarrow{d} cX$

---

## 2. Justification (Why $\boldsymbol{\Sigma}$ can be replaced by $\mathbf{S}$)

### Step 1: Central Limit Theorem (CLT)

We start with the classical multivariate CLT:

$$
\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})
\xrightarrow{d}
\mathcal{N}(\mathbf{0}, \boldsymbol{\Sigma})
$$

---

### Step 2: Standardization

Pre-multiplying by $\boldsymbol{\Sigma}^{-1/2}$ transforms the limiting distribution into a standard normal:

$$
\boldsymbol{\Sigma}^{-1/2}
\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})
\xrightarrow{d}
\mathcal{N}(\mathbf{0}, \mathbf{I})
$$

---

### Step 3: Consistency of the Sample Covariance

By the Law of Large Numbers, the sample covariance matrix satisfies:

$$
\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma}
$$

As a consequence, any continuous matrix function also converges, including:

$$
\mathbf{S}^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}
$$

---

### Step 4: Applying Slutsky’s Theorem

Since $\mathbf{S}^{-1/2}$ converges in probability to a constant matrix, we can multiply it with the CLT result without changing the limiting distribution:

$$
\mathbf{S}^{-1/2}
\left[
\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})
\right]
\xrightarrow{d}
\mathcal{N}(\mathbf{0}, \mathbf{I})
$$

---

### Step 5: Re-scaling Back

Rearranging to restore the original scaling gives:

$$
\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})
\sim
\mathcal{N}(\mathbf{0}, \mathbf{S})
$$

---

### Step 6: Distribution of the Sample Mean

Dividing both sides by $\sqrt{n}$ yields:

$$
\widehat{\boldsymbol{\mu}}_n
\sim
\mathcal{N}\left(\boldsymbol{\mu}, \frac{1}{n}\mathbf{S}\right)
$$

---

# Interpretation

Because the sample covariance $\mathbf{S}$ converges in probability to the true covariance $\boldsymbol{\Sigma}$, replacing $\boldsymbol{\Sigma}$ with $\mathbf{S}$ does not alter the asymptotic distribution. In large samples, both behave equivalently, making the substitution valid under Slutsky’s theorem.

# Part C: Numerical Verification

```python
# @title Numerical Validation (Refactored)
import numpy as np
import pandas as pd
import scipy.stats as stats

# =========================
# Synthetic Strain Data
# =========================
np.random.seed(42)

n_samples = 5000
n_features = 3

mean_vector = [0.5, -0.2, 1.1]
cov_matrix = [
    [0.09, 0.02, 0.01],
    [0.02, 0.06, 0.03],
    [0.01, 0.03, 0.05]
]

# Generate multivariate normal strain data
data = np.random.multivariate_normal(
    mean=mean_vector,
    cov=cov_matrix,
    size=n_samples
)

# Inject subtle structural drift (last 1000 samples)
data[4000:, 0] += 0.015  # Channel 1 drift
data[4000:, 2] -= 0.010  # Channel 3 drift

df_strain = pd.DataFrame(data, columns=["Strain_Ch1", "Strain_Ch2", "Strain_Ch3"])


# =========================
# Statistical Test Function
# =========================
def test_first_moment_homogeneity(df: pd.DataFrame, n_groups: int = 5) -> pd.DataFrame:
    """
    Tests first-moment (mean) homogeneity using:
    - Wilks' Lambda
    - Bartlett's Chi-Square approximation
    """

    n, m = df.shape
    group_size = n // n_groups

    global_mean = df.mean().values

    within_scatter = np.zeros((m, m))
    between_scatter = np.zeros((m, m))

    # =========================
    # Partition into groups
    # =========================
    for i in range(n_groups):
        group = df.iloc[i * group_size : (i + 1) * group_size].values
        group_mean = group.mean(axis=0)

        # Within-group scatter
        centered = group - group_mean
        within_scatter += centered.T @ centered

        # Between-group scatter
        mean_diff = (group_mean - global_mean).reshape(-1, 1)
        between_scatter += group_size * (mean_diff @ mean_diff.T)

    # =========================
    # Wilks' Lambda
    # =========================
    total_scatter = within_scatter + between_scatter
    wilks_lambda = np.linalg.det(within_scatter) / np.linalg.det(total_scatter)

    # =========================
    # Bartlett's Chi-Square
    # =========================
    chi_square = -(n - 1 - (m + n_groups) / 2) * np.log(wilks_lambda)
    degrees_of_freedom = m * (n_groups - 1)

    p_value = stats.chi2.sf(chi_square, degrees_of_freedom)

    # =========================
    # Results Table
    # =========================
    results = pd.DataFrame({
        "Metric": [
            "Wilks' Lambda (Λ)",
            "Bartlett Chi-Square",
            "Degrees of Freedom",
            "p-value"
        ],
        "Value": [
            wilks_lambda,
            chi_square,
            degrees_of_freedom,
            p_value
        ]
    }).set_index("Metric")

    return results


# =========================
# Run Validation
# =========================
results = test_first_moment_homogeneity(df_strain, n_groups=5)
print(results)
```

# **Geometric Subspace Optimization via Principal Component Analysis (PCA)**


# Part A: Coordinate Projections & Orthogonality

## 1)

## Proof of Diagonalization of the Covariance Matrix ($\mathbf{\Lambda}$)

To show that the covariance matrix of the transformed data becomes diagonal, we begin by defining the transformed random vector:

$$
\mathbf{Z}_i = \mathbf{P}^T \widetilde{\mathbf{X}}_i
$$

We then compute its covariance:

$$
\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T]
=
\mathbb{E}\left[(\mathbf{P}^T \widetilde{\mathbf{X}}_i)(\mathbf{P}^T \widetilde{\mathbf{X}}_i)^T\right]
$$

Applying the transpose identity $(AB)^T = B^T A^T$:

$$
=
\mathbb{E}[\mathbf{P}^T \widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T \mathbf{P}]
$$

Since $\mathbf{P}$ is a constant matrix (formed by eigenvectors), it can be taken outside the expectation:

$$
=
\mathbf{P}^T \mathbb{E}[\widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T] \mathbf{P}
$$

Using the definition of the covariance matrix:

$$
\mathbb{E}[\widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T] = \boldsymbol{\Sigma}
$$

we obtain:

$$
=
\mathbf{P}^T \boldsymbol{\Sigma} \mathbf{P}
$$

Substituting the spectral decomposition of $\boldsymbol{\Sigma}$:

$$
\boldsymbol{\Sigma} = \mathbf{P} \mathbf{\Lambda} \mathbf{P}^T
$$

gives:

$$
=
\mathbf{P}^T (\mathbf{P} \mathbf{\Lambda} \mathbf{P}^T) \mathbf{P}
$$

Using orthogonality of $\mathbf{P}$, where $\mathbf{P}^T \mathbf{P} = \mathbf{I}$:

$$
=
(\mathbf{P}^T \mathbf{P}) \mathbf{\Lambda} (\mathbf{P}^T \mathbf{P})
=
\mathbf{I} \mathbf{\Lambda} \mathbf{I}
=
\mathbf{\Lambda}
$$

---

## Geometric and Statistical Interpretation

### Statistical Meaning
A diagonal covariance matrix implies that all off-diagonal terms are zero. This means there is no linear correlation between different components $Z_{i,j}$ and $Z_{i,k}$. The transformed features are therefore fully decorrelated and statistically independent in the second-order sense.

### Geometric Meaning
The transformation rotates the original data space such that the axes align with directions of maximum variance. In this new coordinate system, the data cloud is reoriented along orthogonal principal directions, corresponding to the standard axes ($x, y, z$, etc.), where each axis captures independent variance.

## 2)

## Matrix Trace Proof

To demonstrate that PCA preserves total variance, we show that the trace of the covariance matrix remains unchanged under orthogonal transformation.

The covariance matrix can be decomposed as:

$$
\boldsymbol{\Sigma} = \mathbf{P}\mathbf{\Lambda}\mathbf{P}^T
$$

Taking the trace of both sides:

$$
\text{tr}(\boldsymbol{\Sigma}) = \text{tr}(\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T)
$$

Using the cyclic property of trace, where $\text{tr}(ABC) = \text{tr}(BCA)$:

$$
\text{tr}(\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T)
=
\text{tr}(\mathbf{\Lambda}\mathbf{P}^T\mathbf{P})
$$

Since $\mathbf{P}$ is an orthogonal matrix:

$$
\mathbf{P}^T\mathbf{P} = \mathbf{I}
$$

Thus,

$$
\text{tr}(\mathbf{\Lambda}\mathbf{I}) = \text{tr}(\mathbf{\Lambda})
$$

Because $\mathbf{\Lambda}$ is diagonal with eigenvalues on its diagonal:

$$
\text{tr}(\mathbf{\Lambda}) = \sum_{j=1}^{m} \lambda_j
$$

This confirms that total variance is fully preserved and simply redistributed across principal components.

---

## Variance Ratio Formulations

### Cumulative Explained Variance Ratio $\Phi(k)$

The proportion of total variance retained in the first $k$ principal components is:

$$
\Phi(k) =
\frac{\sum_{j=1}^{k} \widehat{\lambda}_j}
{\sum_{i=1}^{m} \widehat{\lambda}_i}
$$

This measures how much information is preserved in the reduced-dimensional subspace.

---

### Residual (Unexplained) Variance Ratio $\Psi(k)$

The proportion of variance lost by truncating after $k$ components is:

$$
\Psi(k) = 1 - \Phi(k)
$$

Equivalently,

$$
\Psi(k) =
\frac{\sum_{j=k+1}^{m} \widehat{\lambda}_j}
{\sum_{i=1}^{m} \widehat{\lambda}_i}
$$

This represents the amount of discarded variance, often interpreted as noise or reconstruction error.

## 3)

### Definitions:

### Reconstructed Vector ($\widehat{\mathbf{x}}_i$)
The reconstructed observation is obtained by retaining only the top $k$ principal components of the system:

$$
\widehat{\mathbf{x}}_i
=
\widehat{\boldsymbol{\mu}}_n
+
\widehat{\mathbf{P}}_k \mathbf{z}_{i,k}
$$

This represents the low-dimensional approximation of the original signal captured within the dominant subspace.

---

### Residual Error Vector ($\mathbf{e}_i$)
The residual vector represents the information lost due to truncation, corresponding to the discarded subspace:

$$
\mathbf{e}_i
=
\mathbf{x}_i - \widehat{\mathbf{x}}_i
=
\widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}
$$

This term captures unmodeled variability or noise not explained by the retained principal components.

---

## Geometric (Pythagorean) Decomposition

The centered data vector can be decomposed into orthogonal projections onto the retained and residual subspaces:

$$
(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)
=
\widehat{\mathbf{P}}_k \mathbf{z}_{i,k}
+
\widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}
$$

Since the principal subspace ($\widehat{\mathbf{P}}_k$) and residual subspace ($\widehat{\mathbf{P}}_{m-k}$) are orthogonal, their cross-term vanishes. Therefore, the squared norm decomposes according to the Pythagorean theorem:

$$
\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2
=
\|\widehat{\mathbf{P}}_k \mathbf{z}_{i,k}\|^2
+
\|\widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}\|^2
$$

Because orthonormal projection matrices preserve Euclidean norm, this simplifies to:

$$
\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2
=
\|\mathbf{z}_{i,k}\|^2
+
\|\mathbf{z}_{i,m-k}\|^2
$$

---

## Diagnostic Interpretation: $T^2$ vs $Q$ Statistic

### Hotelling’s $T^2$ — In-Subspace Variation (Operational Stress)

Extreme wind loading or similar operational disturbances primarily affect the system within the learned principal subspace. These changes alter the magnitude of existing correlated modes without breaking inter-sensor relationships. The $T^2$ statistic captures this behavior by monitoring variation inside the principal component space.

---

### $Q$ Statistic (Squared Prediction Error) — Out-of-Subspace Faults

Structural damage such as fatigue cracks modifies the underlying physical coupling between sensors. This introduces new patterns of variation that are not represented in the trained subspace, causing energy to accumulate in the residual space. The $Q$ statistic measures this unexplained residual energy, making it sensitive to such model-violating faults.

```python
# @title Synthetic Sensor Dataset Generation
import numpy as np
import pandas as pd

# =========================
# Reproducibility
# =========================
np.random.seed(42)

num_observations = 2500

# =========================
# Latent driving factors
# =========================
primary_mode = np.random.normal(loc=0, scale=1, size=num_observations)
secondary_mode = np.random.normal(loc=0, scale=1, size=num_observations)

# =========================
# Sensor responses
# =========================
sensor_A = 0.85 * primary_mode + 0.10 * secondary_mode + np.random.normal(0, 0.30, num_observations)
sensor_B = 0.80 * primary_mode + 0.15 * secondary_mode + np.random.normal(0, 0.35, num_observations)

sensor_C = 0.12 * primary_mode + 0.90 * secondary_mode + np.random.normal(0, 0.25, num_observations)

sensor_D = 0.02 * primary_mode + 0.05 * secondary_mode + np.random.normal(0, 1.40, num_observations)

# =========================
# Construct DataFrame
# =========================
df_asset = pd.DataFrame(
    np.column_stack([sensor_A, sensor_B, sensor_C, sensor_D]),
    columns=["Sensor_1", "Sensor_2", "Sensor_3", "Sensor_4"]
)

df_asset
```

```python
# @title PCA Validation

import numpy as np
import pandas as pd


def compute_pca_diagnostics(dataframe: pd.DataFrame, columns: list = None):
    """
    Evaluates PCA-based diagnostics:
    - Hotelling's T² statistic (principal subspace)
    - Q statistic / SPE (residual subspace)
    across multiple subspace dimensions k.
    """

    # =========================
    # Feature selection
    # =========================
    if columns is None:
        columns = dataframe.columns.tolist()

    raw_matrix = dataframe[columns].values
    num_samples, num_features = raw_matrix.shape

    # =========================
    # Centering the data
    # =========================
    mean_vector = np.mean(raw_matrix, axis=0)
    centered_matrix = raw_matrix - mean_vector

    # =========================
    # Covariance decomposition
    # =========================
    cov_matrix = np.cov(centered_matrix, rowvar=False, ddof=1)

    eigen_vals, eigen_vecs = np.linalg.eigh(cov_matrix)

    # Sort in descending order
    sorted_indices = np.argsort(eigen_vals)[::-1]
    eigen_vals = eigen_vals[sorted_indices]
    eigen_vecs = eigen_vecs[:, sorted_indices]

    # =========================
    # PCA projection
    # =========================
    score_matrix = centered_matrix @ eigen_vecs

    # Subspace settings
    cutoff_list = [1, 2, 3]

    t2_means = []
    q_means = []

    # =========================
    # Diagnostic computation
    # =========================
    for k in cutoff_list:

        # Principal subspace (T²)
        principal_scores = score_matrix[:, :k]
        principal_eigenvalues = eigen_vals[:k]

        t2_values = np.sum((principal_scores ** 2) / principal_eigenvalues, axis=1)
        t2_means.append(np.mean(t2_values))

        # Residual subspace (Q statistic)
        residual_scores = score_matrix[:, k:]
        q_values = np.sum(residual_scores ** 2, axis=1)
        q_means.append(np.mean(q_values))

    # =========================
    # Results table
    # =========================
    diagnostics_table = pd.DataFrame({
        "PCA Components (k)": cutoff_list,
        "Avg Hotelling T²": t2_means,
        "Avg Q Statistic": q_means
    }).set_index("PCA Components (k)")

    return diagnostics_table


# =========================
# Run diagnostics
# =========================
pca_results = compute_pca_diagnostics(df_asset)
display(pca_results)
```

# **Latent Subspace Decomposition via Factor Analysis (FA)**


# Part A: Theoretical Fundamentals (Generative Model & Commonalities)

## 1).

### Derivation of the Factor Model Decomposition  
We begin with the definition of the correlation matrix:

$$
\mathbf{R} = \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T]
= \mathbb{E}[(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)^T]
$$

---

### Expanding the Expression

$$
=
\boldsymbol{\Lambda}\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T]\boldsymbol{\Lambda}^T
+
\boldsymbol{\Lambda}\mathbb{E}[\mathbf{F}_i\boldsymbol{\epsilon}_i^T]
+
\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T]\boldsymbol{\Lambda}^T
+
\mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T]
$$

---

### Applying Model Assumptions

Using the standard orthogonality and normalization assumptions:

- $\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T] = \mathbf{I}$
- $\mathbb{E}[\mathbf{F}_i\boldsymbol{\epsilon}_i^T] = 0$
- $\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T] = 0$
- $\mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T] = \boldsymbol{\Psi}$

we obtain:

$$
\mathbf{R}
=
\boldsymbol{\Lambda}\mathbf{I}\boldsymbol{\Lambda}^T
+
\mathbf{0}
+
\mathbf{0}
+
\boldsymbol{\Psi}
$$

which simplifies to:

$$
\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}
$$

---

# Interpretation of the Components

### Communality ($h_j^2$)
Represents the portion of variance in a variable that is explained by the common latent factors. In physical systems, this corresponds to shared structural behavior or correlated system-wide dynamics.

---

### Uniqueness ($\varphi_j^2$)
Represents the variance that is specific to each individual variable and not explained by the factor structure. This includes sensor-specific noise, measurement errors, or localized disturbances.

## 2).

## PCA vs. Varimax Rotation

In standard PCA, the extracted components are arranged in a strict order of explained variance. This often leads to mixed signals being combined within the same principal component, making interpretation less clear.

In contrast, Varimax rotation redistributes the factor loadings to encourage a more interpretable structure. It pushes loadings closer to either 0 or 1, effectively separating variables into more distinct and interpretable groups without enforcing a dominance hierarchy among components.

---

## Why Rotation Does Not Change $\mathbf{R}$ or $h_j^2$

Let $\mathbf{T}$ be a rotation matrix satisfying the orthogonality condition:

$$
\mathbf{T}\mathbf{T}^T = \mathbf{I}
$$

If the original loading matrix is $\boldsymbol{\Lambda}$, then the rotated loadings are given by:

$$
\boldsymbol{\Lambda}_{\text{rot}} = \boldsymbol{\Lambda}\mathbf{T}
$$

Now consider the reconstructed covariance structure:

$$
(\boldsymbol{\Lambda}\mathbf{T})(\boldsymbol{\Lambda}\mathbf{T})^T
=
\boldsymbol{\Lambda}\mathbf{T}\mathbf{T}^T\boldsymbol{\Lambda}^T
=
\boldsymbol{\Lambda}\mathbf{I}\boldsymbol{\Lambda}^T
=
\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T
$$

---

## Key Insight

Since the orthogonal rotation cancels out in the reconstruction, the following remain unchanged:

- The correlation structure $\mathbf{R}$
- The communalities $h_j^2$

Only the distribution of variance across components changes. In other words, rotation does not alter the total explained structure—it only redistributes how that structure is expressed across axes, improving interpretability without affecting the underlying statistical content.

## 3).

## Joint Partitioned Covariance Matrix ($\boldsymbol{\Sigma}_{YY}$)

Consider the combined random vector:

$$
\mathbf{Y}_i =
\begin{bmatrix}
\mathbf{F}_i \\
\mathbf{Z}_i
\end{bmatrix}
$$

The full covariance structure can be expressed in block matrix form as:

$$
\boldsymbol{\Sigma}_{YY}
=
\begin{bmatrix}
\mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] & \mathbb{E}[\mathbf{F}_i \mathbf{Z}_i^T] \\
\mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] & \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T]
\end{bmatrix}
$$

Using the known model relationships:

- $\mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] = \mathbf{I}$
- $\mathbb{E}[\mathbf{F}_i \mathbf{Z}_i^T] = \boldsymbol{\Lambda}^T$
- $\mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] = \boldsymbol{\Lambda}$
- $\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{R}$

the covariance matrix simplifies to:

$$
\boldsymbol{\Sigma}_{YY}
=
\begin{bmatrix}
\mathbf{I} & \boldsymbol{\Lambda}^T \\
\boldsymbol{\Lambda} & \mathbf{R}
\end{bmatrix}
$$

---

## Derivation of Thomson’s Estimator

We begin with the conditional expectation formula for multivariate Gaussian variables:

$$
\mathbb{E}[\mathbf{X}_2 \mid \mathbf{x}_1]
=
\boldsymbol{\mu}_2
+
\boldsymbol{\Sigma}_{21}
\boldsymbol{\Sigma}_{11}^{-1}
(\mathbf{x}_1 - \boldsymbol{\mu}_1)
$$

---

## Variable Mapping

We map the general notation to the factor model setting:

- Target variable: $\mathbf{X}_2 = \mathbf{F}_i$
- Observed variable: $\mathbf{x}_1 = \mathbf{Z}_i$
- Means: $\boldsymbol{\mu}_1 = \boldsymbol{\mu}_2 = \mathbf{0}$
- Cross-covariance: $\boldsymbol{\Sigma}_{21} = \boldsymbol{\Lambda}^T$
- Observed covariance: $\boldsymbol{\Sigma}_{11} = \mathbf{R}$

---

## Substitution

Substituting into the conditional expectation formula gives:

$$
\mathbf{f}_i
=
\mathbf{0}
+
\boldsymbol{\Lambda}^T \mathbf{R}^{-1} (\mathbf{z}_i - \mathbf{0})
$$

which simplifies to:

$$
\mathbf{f}_i
=
\boldsymbol{\Lambda}^T \mathbf{R}^{-1} \mathbf{z}_i
$$

```python
# @title run_factor_analysis_engine (Rewritten Version)

import numpy as np
import pandas as pd
from sklearn.decomposition import FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots


def factor_analysis_dashboard(dataframe: pd.DataFrame, n_factors: int):

    # =========================
    # Select numeric features
    # =========================
    numeric_cols = dataframe.select_dtypes(include=[np.number]).columns.tolist()
    if "count" in numeric_cols:
        numeric_cols.remove("count")

    raw_data = dataframe[numeric_cols].dropna().values
    n_samples, n_features = raw_data.shape

    # =========================
    # Dimensional constraints
    # =========================
    if n_samples <= n_features:
        raise ValueError("Invalid: number of samples must exceed number of features.")
    if n_factors >= n_features:
        raise ValueError("Invalid: number of factors must be less than features.")

    # =========================
    # Standardization layer
    # =========================
    mean_vec = np.mean(raw_data, axis=0)
    std_vec = np.std(raw_data, axis=0, ddof=1)
    std_vec[std_vec == 0] = 1e-15  # numerical stability

    standardized_data = (raw_data - mean_vec) / std_vec

    # =========================
    # Factor Analysis Model
    # =========================
    fa_model = FactorAnalysis(
        n_components=n_factors,
        rotation="varimax",
        random_state=42
    )

    fa_model.fit(standardized_data)

    loading_matrix = fa_model.components_.T
    noise_var = fa_model.noise_variance_

    # =========================
    # Structural metrics
    # =========================
    communality_scores = np.sum(loading_matrix ** 2, axis=1)
    uniqueness_scores = noise_var

    factor_scores = fa_model.transform(standardized_data)

    # =========================
    # Console diagnostics
    # =========================
    print("=== Factor Analysis Engine Report ===")
    print(f"Mean Communality: {np.mean(communality_scores) * 100:.2f}%")
    print(f"Mean Uniqueness: {np.mean(uniqueness_scores) * 100:.2f}%")

    sensor_names = numeric_cols
    factor_names = [f"Latent Factor {i+1}" for i in range(n_factors)]

    # =========================
    # Plot layout
    # =========================
    fig = make_subplots(
        rows=2, cols=2,
        horizontal_spacing=0.22,
        vertical_spacing=0.25,
        subplot_titles=(
            "Factor Loadings Magnitude",
            "Variance Split: Communality vs Uniqueness",
            "Sensor Noise Structure (Uniqueness)",
            "Latent Factor Variance Profile"
        )
    )

    # =========================
    # Panel 1: Heatmap (new color scale)
    # =========================
    fig.add_trace(
        go.Heatmap(
            z=np.abs(loading_matrix),
            x=factor_names,
            y=sensor_names,
            colorscale="Viridis",
            showscale=True,
            colorbar=dict(title="Loading Strength")
        ),
        row=1, col=1
    )

    # =========================
    # Panel 2: Stacked bar (new colors)
    # =========================
    fig.add_trace(
        go.Bar(
            y=sensor_names,
            x=communality_scores * 100,
            name="Communality",
            orientation="h",
            marker=dict(color="#4C78A8")  # blue variant
        ),
        row=1, col=2
    )

    fig.add_trace(
        go.Bar(
            y=sensor_names,
            x=uniqueness_scores * 100,
            name="Uniqueness",
            orientation="h",
            marker=dict(color="#F58518")  # orange variant
        ),
        row=1, col=2
    )

    fig.update_layout(barmode="stack")

    # =========================
    # Panel 3: Noise profile (new color)
    # =========================
    fig.add_trace(
        go.Scatter(
            x=sensor_names,
            y=uniqueness_scores,
            mode="lines+markers",
            name="Noise Level",
            line=dict(color="#E45756", width=2),
            marker=dict(symbol="diamond")
        ),
        row=2, col=1
    )

    fig.update_xaxes(tickangle=30, row=2, col=1)

    # =========================
    # Panel 4: Factor variance (new color)
    # =========================
    factor_var = np.var(factor_scores, axis=0, ddof=1)

    fig.add_trace(
        go.Bar(
            x=factor_names,
            y=factor_var,
            name="Factor Variance",
            marker=dict(color="#72B7B2")  # teal variant
        ),
        row=2, col=2
    )

    # =========================
    # Layout styling
    # =========================
    fig.update_layout(
        title="Factor Analysis Subspace Diagnostics Dashboard (Revised)",
        template="plotly_white",
        height=760,
        width=1250,
        legend=dict(orientation="h", x=0.5, xanchor="center", y=1.05),
        margin=dict(l=120, r=80, t=120, b=60)
    )

    fig.show()


# =========================
# Run engine
# =========================
factor_analysis_dashboard(df_asset, n_factors=2)
```

# **Subspace Diagnostics & Feature De-correlation**

# Part A: Develop a Feature Analysis Dashboard

```python
# @title SubspaceDiagnostics (Visually Redesigned Version)

import numpy as np
import pandas as pd
from sklearn.decomposition import PCA, FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots


class LatentSpaceDiagnostics:

    def __init__(self, dataframe: pd.DataFrame):

        self.dataframe = dataframe
        raw = dataframe.values

        self.X = (raw - np.mean(raw, axis=0)) / np.std(raw, axis=0, ddof=1)
        self.n_samples, self.n_features = self.X.shape

    # =========================================================
    # PCA SUITE (Redesigned Styling)
    # =========================================================
    def pca_diagnostics(self):

        pca = PCA()
        pca.fit(self.X)

        scores = pca.transform(self.X)

        fig = make_subplots(
            rows=2, cols=3,
            subplot_titles=(
                "Loadings Structure",
                "Eigenvalue Spectrum",
                "Explained Variance",
                "Residual Energy",
                "T² Trend",
                "Q Trend"
            ),
            horizontal_spacing=0.18,
            vertical_spacing=0.22
        )

        # -------------------------
        # Heatmap (Cividis)
        # -------------------------
        fig.add_trace(
            go.Heatmap(
                z=np.abs(pca.components_.T),
                colorscale="Cividis",
                showscale=False
            ),
            row=1, col=1
        )

        # -------------------------
        # Eigenvalues (purple tone)
        # -------------------------
        fig.add_trace(
            go.Bar(
                y=pca.explained_variance_,
                marker=dict(color="#8E44AD")
            ),
            row=1, col=2
        )

        # -------------------------
        # Explained variance (teal)
        # -------------------------
        fig.add_trace(
            go.Bar(
                y=pca.explained_variance_ratio_ * 100,
                marker=dict(color="#16A085"),
                opacity=0.85
            ),
            row=1, col=3
        )

        fig.add_trace(
            go.Scatter(
                y=np.cumsum(pca.explained_variance_ratio_) * 100,
                mode="lines+markers",
                line=dict(color="#F39C12", width=3, dash="dash"),
                marker=dict(size=6, symbol="circle")
            ),
            row=1, col=3
        )

        # -------------------------
        # Residual variance (gray-blue)
        # -------------------------
        fig.add_trace(
            go.Bar(
                y=100 - np.cumsum(pca.explained_variance_ratio_) * 100,
                marker=dict(color="#95A5A6")
            ),
            row=2, col=1
        )

        # -------------------------
        # T² curve (blue)
        # -------------------------
        k_vals = [1, 2, 3]
        t2_vals, q_vals = [], []

        for k in k_vals:
            t2_vals.append(np.mean(np.sum((scores[:, :k] ** 2) / pca.explained_variance_[:k], axis=1)))
            q_vals.append(np.mean(np.sum(scores[:, k:] ** 2, axis=1)))

        fig.add_trace(
            go.Scatter(
                x=k_vals,
                y=t2_vals,
                mode="lines+markers",
                line=dict(color="#3498DB", width=3),
                marker=dict(symbol="diamond", size=7)
            ),
            row=2, col=2
        )

        # -------------------------
        # Q curve (red)
        # -------------------------
        fig.add_trace(
            go.Scatter(
                x=k_vals,
                y=q_vals,
                mode="lines+markers",
                line=dict(color="#E74C3C", width=3),
                marker=dict(symbol="x", size=8)
            ),
            row=2, col=3
        )

        fig.update_layout(
            title="PCA Diagnostic Suite (Re-styled Visualization)",
            template="plotly_dark",
            height=780,
            width=1250,
            showlegend=False
        )

        fig.show()

    # =========================================================
    # FACTOR ANALYSIS SUITE (Redesigned Styling)
    # =========================================================
    def factor_analysis_diagnostics(self, n_components=2):

        fa = FactorAnalysis(
            n_components=n_components,
            rotation="varimax",
            random_state=42
        )

        fa.fit(self.X)

        loadings = np.abs(fa.components_.T)
        communality = np.sum(loadings ** 2, axis=1)
        uniqueness = fa.noise_variance_

        scores = fa.transform(self.X)

        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=(
                "Factor Loadings Map",
                "Variance Decomposition",
                "Uniqueness Profile",
                "Latent Factor Spread"
            ),
            horizontal_spacing=0.2,
            vertical_spacing=0.25
        )

        # -------------------------
        # Heatmap (Inferno)
        # -------------------------
        fig.add_trace(
            go.Heatmap(
                z=loadings,
                colorscale="Inferno",
                showscale=False
            ),
            row=1, col=1
        )

        # -------------------------
        # Communality (blue)
        # -------------------------
        fig.add_trace(
            go.Bar(
                y=self.dataframe.columns,
                x=communality * 100,
                orientation="h",
                name="Communality",
                marker=dict(color="#2980B9")
            ),
            row=1, col=2
        )

        # -------------------------
        # Uniqueness (orange)
        # -------------------------
        fig.add_trace(
            go.Bar(
                y=self.dataframe.columns,
                x=uniqueness * 100,
                orientation="h",
                name="Uniqueness",
                marker=dict(color="#E67E22")
            ),
            row=1, col=2
        )

        # -------------------------
        # Noise profile (red dashed)
        # -------------------------
        fig.add_trace(
            go.Scatter(
                x=self.dataframe.columns,
                y=uniqueness,
                mode="lines+markers",
                line=dict(color="#C0392B", width=3, dash="dot"),
                marker=dict(symbol="triangle-up", size=7)
            ),
            row=2, col=1
        )

        # -------------------------
        # Factor variance (green)
        # -------------------------
        fig.add_trace(
            go.Bar(
                x=[f"F{i+1}" for i in range(n_components)],
                y=np.var(scores, axis=0),
                marker=dict(color="#27AE60")
            ),
            row=2, col=2
        )

        fig.update_layout(
            title="Factor Analysis Diagnostic Suite (Redesigned)",
            template="plotly_dark",
            height=780,
            width=1150,
            barmode="stack",
            showlegend=False
        )

        fig.show()
```

```python
# @title Synthetic Data Generation (Rewritten Run Block)

import numpy as np
import pandas as pd

# =========================
# Reproducibility setup
# =========================
np.random.seed(42)

sample_count = 2500

# =========================
# Latent factor generation
# =========================
latent_1 = np.random.normal(0, 1, sample_count)
latent_2 = np.random.normal(0, 1, sample_count)

# =========================
# Sensor response model
# =========================
signal_1 = 0.85 * latent_1 + 0.10 * latent_2 + np.random.normal(0, 0.30, sample_count)
signal_2 = 0.80 * latent_1 + 0.15 * latent_2 + np.random.normal(0, 0.35, sample_count)
signal_3 = 0.12 * latent_1 + 0.90 * latent_2 + np.random.normal(0, 0.25, sample_count)
signal_4 = 0.02 * latent_1 + 0.05 * latent_2 + np.random.normal(0, 1.40, sample_count)

# =========================
# DataFrame construction
# =========================
df_system = pd.DataFrame({
    "Channel_A": signal_1,
    "Channel_B": signal_2,
    "Channel_C": signal_3,
    "Channel_D": signal_4
})

# =========================
# Run diagnostics engine
# =========================
diagnostics_engine = LatentSpaceDiagnostics(df_system)
diagnostics_engine.pca_diagnostics()
```

# Q1:

1. **High Uniqueness (~100%) Interpretation**  
   A near-100% uniqueness value for `Sensor_4` indicates that it is largely independent of the rest of the system. In practice, this suggests the sensor is dominated by uncorrelated noise rather than meaningful structural information.

2. **Why PCA Fails Here**  
   PCA prioritizes directions of maximum variance, regardless of whether that variance is informative or noisy. Because `Sensor_4` exhibits extremely high variance, PCA mistakenly treats it as a dominant structural direction and allocates principal components toward modeling noise instead of true system dynamics.

3. **Operational Risk**  
   This leads to unreliable diagnostics: a single noisy sensor can distort the entire PCA model, causing false alarms and masking real structural faults occurring in other parts of the system.

---

# Q2:

1. **PCA vs Factor Analysis (FA)**  
   PCA enforces orthogonal directions that often mix multiple physical effects into the same component. In contrast, Varimax-rotated Factor Analysis redistributes loadings to produce a clearer structure where each factor aligns more closely with a distinct group of sensors.

2. **Interpretability in Practice**  
   With Factor Analysis, a spike in a specific factor can be directly linked to a subset of sensors. PCA components, however, tend to be blended representations, making physical interpretation of failures less straightforward.

---

# Q3:

1. **Behavior of the Q Statistic**  
   The Q-statistic sharply decreases from $k=1$ to $k=2$ as the second dominant structural mode is captured. Beyond $k=2$, the curve flattens, indicating that remaining variance is primarily noise.

2. **Optimal Model Order ($k=2$)**  
   The elbow point at $k=2$ indicates the true dimensionality of the underlying system dynamics. This suggests that two latent factors sufficiently describe the physical behavior.

3. **Overfitting at $k=3$**  
   Selecting $k=3$ introduces unnecessary complexity by modeling noise (including the uninformative behavior of `Sensor_4`) as signal. This leads to overly sensitive and unreliable diagnostic thresholds.

---

# Q4:

1. **PCA Approach**  
   PCA is computationally efficient and simple to implement, but it cannot distinguish between structured signal and high-variance noise, making it vulnerable to sensor corruption.

2. **Factor Analysis Approach**  
   Factor Analysis provides a more robust decomposition by explicitly modeling uniqueness for each sensor. This allows it to identify sensors like `Sensor_4` as noise-dominated and reduce their influence on the latent structure.

3. **Practical Conclusion**  
   Factor Analysis is more suitable for diagnostic systems because it isolates meaningful latent structure while suppressing noise-dominated sensors. PCA, in contrast, can be misled by high-variance but uninformative signals.
